```python
# ==============================================================================
# SYSTEM D'ANALYSE COMPLET D'IMPUTATION - DONNEES NEUROPSYCHOLOGIQUES
# Maladie de Parkinson - NS-Park Dataset
# ==============================================================================
# 
# Auteur: Analyse comparative des methodes d'imputation
# Dataset: NS-Park - Evaluations cognitives longitudinales (n=1407)
# Objectif: Comparer performances sous MAR, MCAR, MNAR avec analyse de sensibilite
# Version: 2.0 - Amelioree
# Date: Octobre 2025
#
# ==============================================================================

# ==============================================================================
# CELLULE 1: IMPORTS ET CONFIGURATION
# ==============================================================================

import os
import sys
import time
import warnings
import json
from datetime import datetime
from pathlib import Path
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Machine Learning & Imputation
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import SimpleImputer, KNNImputer, IterativeImputer
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import BayesianRidge, LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler

# Statistical models
from scipy import stats
from scipy.stats import ks_2samp
import statsmodels.api as sm
try:
    import statsmodels.formula.api as smf
    STATSMODELS_AVAILABLE = True
except:
    STATSMODELS_AVAILABLE = False
    print("Statsmodels non disponible - LMM non utilisable")

# Deep Learning (optionnel)
try:
    import tensorflow as tf
    from tensorflow import keras
    from tensorflow.keras.models import Sequential
    from tensorflow.keras.layers import Dense, LSTM, Bidirectional, Dropout
    TF_AVAILABLE = True
except:
    TF_AVAILABLE = False
    print("TensorFlow non disponible - methodes DL non utilisables")

# HyperImpute
try:
    from hyperimpute.plugins.imputers import Imputers
    HYPERIMPUTE_AVAILABLE = True
except:
    HYPERIMPUTE_AVAILABLE = False
    print("HyperImpute non disponible - installer via: pip install hyperimpute")

# VaDER
try:
    VADER_PATH = "/content/VaDER"
    if os.path.exists(VADER_PATH):
        sys.path.append(VADER_PATH)
        sys.path.append(os.path.join(VADER_PATH, "VaDER"))
        from vader import VADER
        VADER_AVAILABLE = True
    else:
        VADER_AVAILABLE = False
        print("VaDER non disponible - cloner depuis: git clone https://github.com/johanndejong/VaDER.git")
except:
    VADER_AVAILABLE = False
    print("VaDER import error")

# Configuration visualisation
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)

print("=" * 80)
print("SYSTEME D'ANALYSE D'IMPUTATION - INITIALISE")
print("=" * 80)
print(f"TensorFlow disponible: {TF_AVAILABLE}")
print(f"HyperImpute disponible: {HYPERIMPUTE_AVAILABLE}")
print(f"VaDER disponible: {VADER_AVAILABLE}")
print(f"Statsmodels disponible: {STATSMODELS_AVAILABLE}")

# ==============================================================================
# CELLULE 2: CHARGEMENT ET PREPARATION DES DONNEES
# ==============================================================================

def load_and_prepare_data(filepath):
    """
    Charge et prepare le dataset neuropsychologique NS-Park
    
    Parameters:
        filepath: Chemin vers le fichier Excel
        
    Returns:
        df_numeric: DataFrame avec variables numeriques
        df_original: DataFrame original
        metadata: Metadonnees
    """
    print("\n" + "=" * 80)
    print("CHARGEMENT DES DONNEES")
    print("=" * 80)
    
    # Chargement
    df = pd.read_excel(filepath, sheet_name='NS-Park')
    print(f"Dataset charge: {df.shape[0]} observations, {df.shape[1]} variables")
    
    # Identification des colonnes
    id_cols = ['Identifiant', 'date exam', 'date naiss']
    demographic_cols = ['Age', 'Sexe', 'dur.Sco']
    
    # Variables cognitives
    cognitive_cols = [col for col in df.columns if col not in id_cols + demographic_cols]
    
    # Conversion types
    df_work = df.copy()
    df_work['Age'] = pd.to_numeric(df_work['Age'], errors='coerce')
    df_work['dur.Sco'] = pd.to_numeric(df_work['dur.Sco'], errors='coerce')
    df_work['Sexe_binary'] = (df_work['Sexe'] == 'Homme').astype(int)
    
    # Dataset numerique
    numeric_cols = ['Age', 'dur.Sco', 'Sexe_binary'] + cognitive_cols
    df_numeric = df_work[numeric_cols].copy()
    
    # Nettoyage valeurs manquantes initiales
    initial_missing = df_numeric.isnull().sum().sum()
    print(f"Valeurs manquantes initiales: {initial_missing}")
    
    for col in df_numeric.columns:
        if df_numeric[col].isnull().any():
            df_numeric[col].fillna(df_numeric[col].median(), inplace=True)
            
    print(f"Dataset numerique prepare: {df_numeric.shape}")
    print(f"  - Variables demographiques: 3")
    print(f"  - Variables cognitives: {len(cognitive_cols)}")
    
    # Statistiques descriptives
    print("\n" + "-" * 80)
    print("STATISTIQUES DESCRIPTIVES")
    print("-" * 80)
    print(df_numeric.describe().T[['mean', 'std', 'min', 'max']])
    
    # Metadonnees
    metadata = {
        'n_obs': len(df_numeric),
        'n_vars': len(df_numeric.columns),
        'cognitive_vars': cognitive_cols,
        'demographic_vars': ['Age', 'dur.Sco', 'Sexe_binary'],
        'original_data': df
    }
    
    return df_numeric, df, metadata


# ==============================================================================
# CELLULE 3: SIMULATION DES DONNEES MANQUANTES (MAR, MCAR, MNAR)
# ==============================================================================

def simulate_missing_data(df, missing_rate=0.20, mechanism='MAR', seed=42):
    """
    Simule des donnees manquantes selon differents mecanismes
    
    Parameters:
        df: DataFrame complet
        missing_rate: Proportion de donnees manquantes (0-1)
        mechanism: 'MCAR', 'MAR', ou 'MNAR'
        seed: Graine aleatoire
        
    Returns:
        df_missing: DataFrame avec donnees manquantes
        mask_missing: Masque booleen (True = manquant)
        ground_truth: Valeurs originales
    """
    np.random.seed(seed)
    df_complete = df.copy()
    df_missing = df.copy()
    
    n_rows, n_cols = df.shape
    cognitive_cols = [col for col in df.columns if col not in ['Age', 'dur.Sco', 'Sexe_binary']]
    
    print(f"\n{'=' * 80}")
    print(f"SIMULATION: {mechanism} - Taux cible: {missing_rate:.1%}")
    print(f"{'=' * 80}")
    
    if mechanism == 'MCAR':
        # Missing Completely At Random
        mask = np.random.random((n_rows, len(cognitive_cols))) < missing_rate
        
        for idx, col in enumerate(cognitive_cols):
            df_missing.loc[mask[:, idx], col] = np.nan
            
        print(f"MCAR: Missingness aleatoire uniforme")
        
    elif mechanism == 'MAR':
        # Missing At Random - dependant de l'age
        age_normalized = (df['Age'] - df['Age'].min()) / (df['Age'].max() - df['Age'].min())
        base_prob = missing_rate * 0.5
        
        for col in cognitive_cols:
            prob_missing = base_prob + (missing_rate * age_normalized)
            prob_missing = np.clip(prob_missing, 0, 0.5)
            
            mask = np.random.random(n_rows) < prob_missing
            df_missing.loc[mask, col] = np.nan
            
        print(f"MAR: Missingness dependant de l'age (correlation positive)")
        
    elif mechanism == 'MNAR':
        # Missing Not At Random - dependant de la valeur elle-meme
        for col in cognitive_cols:
            col_values = df[col].values
            col_normalized = (col_values - col_values.min()) / (col_values.max() - col_values.min() + 1e-10)
            
            # Probabilite inversement proportionnelle a la valeur
            prob_missing = missing_rate * (1 - col_normalized)
            prob_missing = np.clip(prob_missing, missing_rate * 0.3, missing_rate * 1.5)
            
            mask = np.random.random(n_rows) < prob_missing
            df_missing.loc[mask, col] = np.nan
            
        print(f"MNAR: Missingness dependant des valeurs (scores bas plus manquants)")
    
    # Calcul statistiques
    mask_missing = df_missing.isnull()
    n_missing = mask_missing.sum().sum()
    actual_rate = n_missing / (n_rows * len(cognitive_cols))
    
    print(f"\nSTATISTIQUES DE MISSINGNESS:")
    print(f"  - Cellules manquantes: {n_missing:,}")
    print(f"  - Taux reel: {actual_rate:.2%}")
    print(f"  - Variables affectees: {mask_missing.any().sum()}/{len(cognitive_cols)}")
    
    # Statistiques par variable
    missing_per_var = mask_missing.sum()
    print(f"\n  Missingness par variable (top 10):")
    top_missing = missing_per_var[missing_per_var > 0].sort_values(ascending=False).head(10)
    for var, count in top_missing.items():
        pct = 100 * count / n_rows
        print(f"    {var:20s}: {count:4d} ({pct:5.1f}%)")
    
    ground_truth = df_complete.copy()
    
    return df_missing, mask_missing, ground_truth


# ==============================================================================
# CELLULE 4: CLASSES D'IMPUTATION
# ==============================================================================

class ImputationMethod:
    """Classe de base pour toutes les methodes d'imputation"""
    
    def __init__(self, name, method_type):
        self.name = name
        self.method_type = method_type
        self.execution_time = None
        self.hyperparameters = {}
        
    def fit_transform(self, X_train, X_test=None):
        raise NotImplementedError
        
    def get_params(self):
        return self.hyperparameters


class MeanImputerMethod(ImputationMethod):
    """Imputation par moyenne"""
    
    def __init__(self):
        super().__init__("Mean", "simple")
        self.hyperparameters = {'strategy': 'mean'}
        self.imputer = None
        
    def fit_transform(self, X_train, X_test=None):
        start_time = datetime.now()
        
        self.imputer = SimpleImputer(strategy='mean')
        X_train_imputed = pd.DataFrame(
            self.imputer.fit_transform(X_train),
            columns=X_train.columns,
            index=X_train.index
        )
        
        if X_test is not None:
            X_test_imputed = pd.DataFrame(
                self.imputer.transform(X_test),
                columns=X_test.columns,
                index=X_test.index
            )
        else:
            X_test_imputed = None
            
        self.execution_time = (datetime.now() - start_time).total_seconds()
        
        return X_train_imputed, X_test_imputed


class MedianImputerMethod(ImputationMethod):
    """Imputation par mediane"""
    
    def __init__(self):
        super().__init__("Median", "simple")
        self.hyperparameters = {'strategy': 'median'}
        self.imputer = None
        
    def fit_transform(self, X_train, X_test=None):
        start_time = datetime.now()
        
        self.imputer = SimpleImputer(strategy='median')
        X_train_imputed = pd.DataFrame(
            self.imputer.fit_transform(X_train),
            columns=X_train.columns,
            index=X_train.index
        )
        
        if X_test is not None:
            X_test_imputed = pd.DataFrame(
                self.imputer.transform(X_test),
                columns=X_test.columns,
                index=X_test.index
            )
        else:
            X_test_imputed = None
            
        self.execution_time = (datetime.now() - start_time).total_seconds()
        
        return X_train_imputed, X_test_imputed


class KNNImputerMethod(ImputationMethod):
    """Imputation par K-Nearest Neighbors"""
    
    def __init__(self, n_neighbors=5):
        super().__init__("KNN", "advanced")
        self.n_neighbors = n_neighbors
        self.hyperparameters = {
            'n_neighbors': n_neighbors,
            'weights': 'uniform'
        }
        self.imputer = None
        
    def fit_transform(self, X_train, X_test=None):
        start_time = datetime.now()
        
        self.imputer = KNNImputer(n_neighbors=self.n_neighbors, weights='uniform')
        X_train_imputed = pd.DataFrame(
            self.imputer.fit_transform(X_train),
            columns=X_train.columns,
            index=X_train.index
        )
        
        if X_test is not None:
            X_test_imputed = pd.DataFrame(
                self.imputer.transform(X_test),
                columns=X_test.columns,
                index=X_test.index
            )
        else:
            X_test_imputed = None
            
        self.execution_time = (datetime.now() - start_time).total_seconds()
        
        return X_train_imputed, X_test_imputed


class MICEImputerMethod(ImputationMethod):
    """Multiple Imputation by Chained Equations"""
    
    def __init__(self, max_iter=10, n_nearest_features=10):
        super().__init__("MICE", "advanced")
        self.max_iter = max_iter
        self.n_nearest_features = n_nearest_features
        self.hyperparameters = {
            'max_iter': max_iter,
            'n_nearest_features': n_nearest_features,
            'estimator': 'BayesianRidge'
        }
        self.imputer = None
        
    def fit_transform(self, X_train, X_test=None):
        start_time = datetime.now()
        
        self.imputer = IterativeImputer(
            max_iter=self.max_iter,
            n_nearest_features=self.n_nearest_features,
            random_state=42
        )
        
        X_train_imputed = pd.DataFrame(
            self.imputer.fit_transform(X_train),
            columns=X_train.columns,
            index=X_train.index
        )
        
        if X_test is not None:
            X_test_imputed = pd.DataFrame(
                self.imputer.transform(X_test),
                columns=X_test.columns,
                index=X_test.index
            )
        else:
            X_test_imputed = None
            
        self.execution_time = (datetime.now() - start_time).total_seconds()
        
        return X_train_imputed, X_test_imputed


class MissForestImputerMethod(ImputationMethod):
    """Imputation par Random Forest (MissForest)"""
    
    def __init__(self, n_estimators=100, max_iter=10):
        super().__init__("MissForest", "ml")
        self.n_estimators = n_estimators
        self.max_iter = max_iter
        self.hyperparameters = {
            'n_estimators': n_estimators,
            'max_iter': max_iter,
            'random_state': 42
        }
        self.imputer = None
        
    def fit_transform(self, X_train, X_test=None):
        start_time = datetime.now()
        
        self.imputer = IterativeImputer(
            estimator=RandomForestRegressor(n_estimators=self.n_estimators, random_state=42),
            max_iter=self.max_iter,
            random_state=42
        )
        
        X_train_imputed = pd.DataFrame(
            self.imputer.fit_transform(X_train),
            columns=X_train.columns,
            index=X_train.index
        )
        
        if X_test is not None:
            X_test_imputed = pd.DataFrame(
                self.imputer.transform(X_test),
                columns=X_test.columns,
                index=X_test.index
            )
        else:
            X_test_imputed = None
            
        self.execution_time = (datetime.now() - start_time).total_seconds()
        
        return X_train_imputed, X_test_imputed


class EMImputerMethod(ImputationMethod):
    """Expectation-Maximization Algorithm"""
    
    def __init__(self, max_iter=100, tol=1e-3):
        super().__init__("EM", "advanced")
        self.max_iter = max_iter
        self.tol = tol
        self.hyperparameters = {
            'max_iter': max_iter,
            'tolerance': tol
        }
        self.mean_ = None
        self.cov_ = None
        
    def fit_transform(self, X_train, X_test=None):
        start_time = datetime.now()
        
        X_imputed = X_train.copy()
        
        # Initialisation avec la moyenne
        for col in X_imputed.columns:
            mask = X_imputed[col].isnull()
            if mask.any():
                X_imputed.loc[mask, col] = X_train[col].mean()
        
        # Iterations EM
        for iteration in range(self.max_iter):
            X_old = X_imputed.copy()
            
            # E-step
            self.mean_ = X_imputed.mean()
            self.cov_ = X_imputed.cov()
            
            # M-step
            for col in X_train.columns:
                mask = X_train[col].isnull()
                if mask.any():
                    other_cols = [c for c in X_train.columns if c != col]
                    X_obs = X_imputed.loc[~mask, other_cols]
                    y_obs = X_imputed.loc[~mask, col]
                    X_miss = X_imputed.loc[mask, other_cols]
                    
                    if len(X_obs) > 0:
                        model = LinearRegression()
                        model.fit(X_obs, y_obs)
                        y_pred = model.predict(X_miss)
                        X_imputed.loc[mask, col] = y_pred
            
            # Convergence
            diff = np.abs(X_imputed - X_old).mean().mean()
            if diff < self.tol:
                break
        
        X_train_imputed = X_imputed
        
        if X_test is not None:
            X_test_imputed = X_test.copy()
            for col in X_test.columns:
                mask = X_test[col].isnull()
                if mask.any() and col in self.mean_.index:
                    X_test_imputed.loc[mask, col] = self.mean_[col]
        else:
            X_test_imputed = None
            
        self.execution_time = (datetime.now() - start_time).total_seconds()
        
        return X_train_imputed, X_test_imputed


class HyperImputeMethod(ImputationMethod):
    """HyperImpute - Ensemble adaptatif avec optimisation automatique"""
    
    def __init__(self):
        super().__init__("HyperImpute", "ensemble")
        self.hyperparameters = {
            'method': 'hyperimpute',
            'optimizer': 'auto'
        }
        self.plugin = None
        
    def fit_transform(self, X_train, X_test=None):
        if not HYPERIMPUTE_AVAILABLE:
            raise ImportError("HyperImpute non disponible")
            
        start_time = datetime.now()
        
        try:
            imputers = Imputers()
            self.plugin = imputers.get("hyperimpute")
            
            # Fit sur train
            X_train_imputed = self.plugin.fit_transform(X_train.copy())
            
            # Transform test
            if X_test is not None:
                X_test_imputed = self.plugin.transform(X_test.copy())
            else:
                X_test_imputed = None
                
            # Conversion en DataFrame si necessaire
            if not isinstance(X_train_imputed, pd.DataFrame):
                X_train_imputed = pd.DataFrame(
                    X_train_imputed, 
                    columns=X_train.columns, 
                    index=X_train.index
                )
            
            if X_test_imputed is not None and not isinstance(X_test_imputed, pd.DataFrame):
                X_test_imputed = pd.DataFrame(
                    X_test_imputed,
                    columns=X_test.columns,
                    index=X_test.index
                )
            
        except Exception as e:
            print(f"Erreur HyperImpute: {e}")
            # Fallback sur MICE
            mice = MICEImputerMethod()
            return mice.fit_transform(X_train, X_test)
            
        self.execution_time = (datetime.now() - start_time).total_seconds()
        
        return X_train_imputed, X_test_imputed


class VaDERImputerMethod(ImputationMethod):
    """VaDER - Variational Autoencoder with Deep Embedding Regularization"""
    
    def __init__(self, n_hidden=[128, 32], learning_rate=1e-5, n_epochs=100):
        super().__init__("VaDER", "deep_learning")
        self.n_hidden = n_hidden
        self.learning_rate = learning_rate
        self.n_epochs = n_epochs
        self.hyperparameters = {
            'n_hidden': n_hidden,
            'learning_rate': learning_rate,
            'n_epochs': n_epochs,
            'cell_type': 'LSTM'
        }
        self.vader_model = None
        
    def fit_transform(self, X_train, X_test=None):
        if not VADER_AVAILABLE:
            raise ImportError("VaDER non disponible")
            
        start_time = datetime.now()
        
        try:
            # Preparation des donnees
            X_train_array = X_train.values
            W_train = np.isnan(X_train_array).astype(int)
            
            X_train_reshaped = X_train_array.reshape(X_train_array.shape[0], 1, X_train_array.shape[1])
            W_train_reshaped = W_train.reshape(W_train.shape[0], 1, W_train.shape[1])
            
            # Initialisation VaDER
            self.vader_model = VADER(
                X_train=X_train_reshaped,
                n_hidden=self.n_hidden,
                k=3,
                learning_rate=self.learning_rate,
                batch_size=min(64, len(X_train) // 4),
                output_activation=None,
                recurrent=True,
                cell_type='LSTM',
                W_train=W_train_reshaped,
                seed=42
            )
            
            # Reconstruction
            X_reconstructed_train = self.vader_model.predict(X_train_reshaped, W_train_reshaped)
            X_reconstructed_train_reshaped = X_reconstructed_train.reshape(X_train_array.shape)
            
            # Imputation
            X_train_imputed = X_train.copy()
            X_train_imputed_array = X_train_array.copy()
            X_train_imputed_array[W_train == 1] = X_reconstructed_train_reshaped[W_train == 1]
            X_train_imputed.loc[:, :] = X_train_imputed_array
            
            # Test
            if X_test is not None:
                X_test_array = X_test.values
                W_test = np.isnan(X_test_array).astype(int)
                
                X_test_reshaped = X_test_array.reshape(X_test_array.shape[0], 1, X_test_array.shape[1])
                W_test_reshaped = W_test.reshape(W_test.shape[0], 1, W_test.shape[1])
                
                X_reconstructed_test = self.vader_model.predict(X_test_reshaped, W_test_reshaped)
                X_reconstructed_test_reshaped = X_reconstructed_test.reshape(X_test_array.shape)
                
                X_test_imputed = X_test.copy()
                X_test_imputed_array = X_test_array.copy()
                X_test_imputed_array[W_test == 1] = X_reconstructed_test_reshaped[W_test == 1]
                X_test_imputed.loc[:, :] = X_test_imputed_array
            else:
                X_test_imputed = None
                
        except Exception as e:
            print(f"Erreur VaDER: {e}")
            # Fallback sur MICE
            mice = MICEImputerMethod()
            return mice.fit_transform(X_train, X_test)
            
        self.execution_time = (datetime.now() - start_time).total_seconds()
        
        return X_train_imputed, X_test_imputed


class LSTMImputerMethod(ImputationMethod):
    """LSTM pour imputation de series temporelles"""
    
    def __init__(self, lstm_units=64, epochs=50):
        super().__init__("LSTM", "deep_learning")
        self.lstm_units = lstm_units
        self.epochs = epochs
        self.hyperparameters = {
            'lstm_units': lstm_units,
            'epochs': epochs,
            'optimizer': 'adam'
        }
        self.model = None
        
    def fit_transform(self, X_train, X_test=None):
        if not TF_AVAILABLE:
            raise ImportError("TensorFlow requis pour LSTM")
            
        start_time = datetime.now()
        
        try:
            # Imputation simple initiale
            simple_imputer = SimpleImputer(strategy='mean')
            X_train_filled = pd.DataFrame(
                simple_imputer.fit_transform(X_train),
                columns=X_train.columns,
                index=X_train.index
            )
            
            # Build LSTM model
            n_features = X_train.shape[1]
            
            model = Sequential([
                LSTM(self.lstm_units, return_sequences=True, input_shape=(1, n_features)),
                Dropout(0.2),
                LSTM(self.lstm_units // 2, return_sequences=False),
                Dropout(0.2),
                Dense(n_features)
            ])
            
            model.compile(optimizer='adam', loss='mse')
            
            # Preparation donnees
            X_train_reshaped = X_train_filled.values.reshape(-1, 1, n_features)
            
            # Entrainement
            model.fit(X_train_reshaped, X_train_filled.values, 
                     epochs=self.epochs, batch_size=32, verbose=0)
            
            # Prediction
            X_train_pred = model.predict(X_train_reshaped, verbose=0)
            X_train_imputed = pd.DataFrame(
                X_train_pred,
                columns=X_train.columns,
                index=X_train.index
            )
            
            # Test
            if X_test is not None:
                X_test_filled = pd.DataFrame(
                    simple_imputer.transform(X_test),
                    columns=X_test.columns,
                    index=X_test.index
                )
                X_test_reshaped = X_test_filled.values.reshape(-1, 1, n_features)
                X_test_pred = model.predict(X_test_reshaped, verbose=0)
                X_test_imputed = pd.DataFrame(
                    X_test_pred,
                    columns=X_test.columns,
                    index=X_test.index
                )
            else:
                X_test_imputed = None
                
        except Exception as e:
            print(f"Erreur LSTM: {e}")
            mice = MICEImputerMethod()
            return mice.fit_transform(X_train, X_test)
            
        self.execution_time = (datetime.now() - start_time).total_seconds()
        
        return X_train_imputed, X_test_imputed


class BiLSTMImputerMethod(ImputationMethod):
    """Bidirectional LSTM pour imputation"""
    
    def __init__(self, lstm_units=64, epochs=50):
        super().__init__("Bi-LSTM", "deep_learning")
        self.lstm_units = lstm_units
        self.epochs = epochs
        self.hyperparameters = {
            'lstm_units': lstm_units,
            'epochs': epochs,
            'bidirectional': True
        }
        
    def fit_transform(self, X_train, X_test=None):
        if not TF_AVAILABLE:
            raise ImportError("TensorFlow requis pour Bi-LSTM")
            
        start_time = datetime.now()
        
        try:
            simple_imputer = SimpleImputer(strategy='mean')
            X_train_filled = pd.DataFrame(
                simple_imputer.fit_transform(X_train),
                columns=X_train.columns,
                index=X_train.index
            )
            
            n_features = X_train.shape[1]
            
            model = Sequential([
                Bidirectional(LSTM(self.lstm_units, return_sequences=True), 
                            input_shape=(1, n_features)),
                Dropout(0.2),
                Bidirectional(LSTM(self.lstm_units // 2, return_sequences=False)),
                Dropout(0.2),
                Dense(n_features)
            ])
            
            model.compile(optimizer='adam', loss='mse')
            
            X_train_reshaped = X_train_filled.values.reshape(-1, 1, n_features)
            model.fit(X_train_reshaped, X_train_filled.values,
                     epochs=self.epochs, batch_size=32, verbose=0)
            
            X_train_pred = model.predict(X_train_reshaped, verbose=0)
            X_train_imputed = pd.DataFrame(
                X_train_pred,
                columns=X_train.columns,
                index=X_train.index
            )
            
            if X_test is not None:
                X_test_filled = pd.DataFrame(
                    simple_imputer.transform(X_test),
                    columns=X_test.columns,
                    index=X_test.index
                )
                X_test_reshaped = X_test_filled.values.reshape(-1, 1, n_features)
                X_test_pred = model.predict(X_test_reshaped, verbose=0)
                X_test_imputed = pd.DataFrame(
                    X_test_pred,
                    columns=X_test.columns,
                    index=X_test.index
                )
            else:
                X_test_imputed = None
                
        except Exception as e:
            print(f"Erreur Bi-LSTM: {e}")
            mice = MICEImputerMethod()
            return mice.fit_transform(X_train, X_test)
            
        self.execution_time = (datetime.now() - start_time).total_seconds()
        
        return X_train_imputed, X_test_imputed


class DNNImputerMethod(ImputationMethod):
    """Deep Neural Network pour imputation"""
    
    def __init__(self, hidden_layers=[128, 64, 32], epochs=100):
        super().__init__("DNN", "deep_learning")
        self.hidden_layers = hidden_layers
        self.epochs = epochs
        self.hyperparameters = {
            'hidden_layers': hidden_layers,
            'epochs': epochs,
            'activation': 'relu'
        }
        
    def fit_transform(self, X_train, X_test=None):
        if not TF_AVAILABLE:
            raise ImportError("TensorFlow requis pour DNN")
            
        start_time = datetime.now()
        
        try:
            simple_imputer = SimpleImputer(strategy='mean')
            X_train_filled = pd.DataFrame(
                simple_imputer.fit_transform(X_train),
                columns=X_train.columns,
                index=X_train.index
            )
            
            n_features = X_train.shape[1]
            
            model = Sequential()
            model.add(Dense(self.hidden_layers[0], activation='relu', input_dim=n_features))
            model.add(Dropout(0.3))
            
            for units in self.hidden_layers[1:]:
                model.add(Dense(units, activation='relu'))
                model.add(Dropout(0.3))
            
            model.add(Dense(n_features))
            
            model.compile(optimizer='adam', loss='mse')
            
            model.fit(X_train_filled.values, X_train_filled.values,
                     epochs=self.epochs, batch_size=32, verbose=0)
            
            X_train_pred = model.predict(X_train_filled.values, verbose=0)
            X_train_imputed = pd.DataFrame(
                X_train_pred,
                columns=X_train.columns,
                index=X_train.index
            )
            
            if X_test is not None:
                X_test_filled = pd.DataFrame(
                    simple_imputer.transform(X_test),
                    columns=X_test.columns,
                    index=X_test.index
                )
                X_test_pred = model.predict(X_test_filled.values, verbose=0)
                X_test_imputed = pd.DataFrame(
                    X_test_pred,
                    columns=X_test.columns,
                    index=X_test.index
                )
            else:
                X_test_imputed = None
                
        except Exception as e:
            print(f"Erreur DNN: {e}")
            mice = MICEImputerMethod()
            return mice.fit_transform(X_train, X_test)
            
        self.execution_time = (datetime.now() - start_time).total_seconds()
        
        return X_train_imputed, X_test_imputed


# ==============================================================================
# CELLULE 5: METRIQUES D'EVALUATION
# ==============================================================================

def calculate_imputation_metrics(y_true, y_pred, mask):
    """
    Calcule les metriques d'evaluation de l'imputation
    
    Parameters:
        y_true: Valeurs reelles
        y_pred: Valeurs imputees
        mask: Masque des valeurs manquantes
        
    Returns:
        dict: Dictionnaire des metriques
    """
    true_values = []
    pred_values = []
    
    for col in y_true.columns:
        if col in y_pred.columns and col in mask.columns:
            missing_idx = mask[col]
            if missing_idx.any():
                true_vals = y_true.loc[missing_idx, col].values
                pred_vals = y_pred.loc[missing_idx, col].values
                
                valid_mask = ~(np.isnan(true_vals) | np.isnan(pred_vals))
                true_values.extend(true_vals[valid_mask])
                pred_values.extend(pred_vals[valid_mask])
    
    if len(true_values) == 0:
        return {
            'MAE': np.nan,
            'RMSE': np.nan,
            'MSE': np.nan,
            'R2': np.nan,
            'MAPE': np.nan,
            'Correlation': np.nan
        }
    
    true_values = np.array(true_values)
    pred_values = np.array(pred_values)
    
    mae = mean_absolute_error(true_values, pred_values)
    mse = mean_squared_error(true_values, pred_values)
    rmse = np.sqrt(mse)
    
    try:
        r2 = r2_score(true_values, pred_values)
    except:
        r2 = np.nan
    
    mape = np.mean(np.abs((true_values - pred_values) / (true_values + 1e-10))) * 100
    
    try:
        correlation = np.corrcoef(true_values, pred_values)[0, 1]
    except:
        correlation = np.nan
    
    metrics = {
        'MAE': mae,
        'RMSE': rmse,
        'MSE': mse,
        'R2': r2,
        'MAPE': mape,
        'Correlation': correlation
    }
    
    return metrics


def calculate_distributional_metrics(y_true, y_pred, mask):
    """
    Calcule les metriques de similarite des distributions
    """
    ks_statistics = []
    ks_pvalues = []
    
    for col in y_true.columns:
        if col in y_pred.columns and col in mask.columns:
            missing_idx = mask[col]
            if missing_idx.any():
                true_vals = y_true.loc[missing_idx, col].dropna()
                pred_vals = y_pred.loc[missing_idx, col].dropna()
                
                if len(true_vals) > 5 and len(pred_vals) > 5:
                    ks_stat, ks_pval = ks_2samp(true_vals, pred_vals)
                    ks_statistics.append(ks_stat)
                    ks_pvalues.append(ks_pval)
    
    if len(ks_statistics) == 0:
        return {
            'KS_statistic_mean': np.nan,
            'KS_pvalue_mean': np.nan,
            'Distribution_similarity': np.nan
        }
    
    return {
        'KS_statistic_mean': np.mean(ks_statistics),
        'KS_pvalue_mean': np.mean(ks_pvalues),
        'Distribution_similarity': 1 - np.mean(ks_statistics)
    }


# ==============================================================================
# CELLULE 6: PIPELINE D'EVALUATION
# ==============================================================================

def evaluate_imputation_pipeline(df_complete, missing_mechanism, missing_rate=0.20,
                                 test_size=0.2, n_iterations=5, seed=42):
    """
    Pipeline complet d'evaluation des methodes d'imputation
    """
    print(f"\n{'=' * 80}")
    print(f"PIPELINE D'EVALUATION: {missing_mechanism}")
    print(f"{'=' * 80}")
    
    # Initialisation des methodes
    methods = [
        MeanImputerMethod(),
        MedianImputerMethod(),
        KNNImputerMethod(n_neighbors=5),
        MICEImputerMethod(max_iter=10, n_nearest_features=10),
        MissForestImputerMethod(n_estimators=100, max_iter=5),
        EMImputerMethod(max_iter=50, tol=1e-3)
    ]
    
    # Ajout methodes avancees si disponibles
    if HYPERIMPUTE_AVAILABLE:
        methods.append(HyperImputeMethod())
    
    if VADER_AVAILABLE:
        methods.append(VaDERImputerMethod(n_hidden=[128, 32], n_epochs=50))
    
    if TF_AVAILABLE:
        methods.extend([
            DNNImputerMethod(hidden_layers=[128, 64, 32], epochs=50),
            LSTMImputerMethod(lstm_units=64, epochs=30),
            BiLSTMImputerMethod(lstm_units=64, epochs=30)
        ])
    
    all_results = []
    detailed_results = {}
    
    for iteration in range(n_iterations):
        print(f"\n{'-' * 80}")
        print(f"Iteration {iteration + 1}/{n_iterations}")
        print(f"{'-' * 80}")
        
        # Simulation donnees manquantes
        df_missing, mask_missing, ground_truth = simulate_missing_data(
            df_complete,
            missing_rate=missing_rate,
            mechanism=missing_mechanism,
            seed=seed + iteration
        )
        
        # Split train/test
        train_idx, test_idx = train_test_split(
            np.arange(len(df_missing)),
            test_size=test_size,
            random_state=seed + iteration
        )
        
        X_train = df_missing.iloc[train_idx].copy()
        X_test = df_missing.iloc[test_idx].copy()
        
        X_train_true = ground_truth.iloc[train_idx].copy()
        X_test_true = ground_truth.iloc[test_idx].copy()
        
        mask_train = mask_missing.iloc[train_idx].copy()
        mask_test = mask_missing.iloc[test_idx].copy()
        
        # Evaluation de chaque methode
        for method in methods:
            print(f"\n  Evaluation: {method.name}...", end=' ')
            
            try:
                # Imputation
                X_train_imputed, X_test_imputed = method.fit_transform(X_train, X_test)
                
                # Metriques train
                metrics_train = calculate_imputation_metrics(X_train_true, X_train_imputed, mask_train)
                dist_metrics_train = calculate_distributional_metrics(X_train_true, X_train_imputed, mask_train)
                
                # Metriques test
                metrics_test = calculate_imputation_metrics(X_test_true, X_test_imputed, mask_test)
                dist_metrics_test = calculate_distributional_metrics(X_test_true, X_test_imputed, mask_test)
                
                # Stockage resultats
                result = {
                    'Mechanism': missing_mechanism,
                    'Method': method.name,
                    'Iteration': iteration + 1,
                    'Execution_Time': method.execution_time,
                    
                    # Metriques Train
                    'Train_MAE': metrics_train['MAE'],
                    'Train_RMSE': metrics_train['RMSE'],
                    'Train_MSE': metrics_train['MSE'],
                    'Train_R2': metrics_train['R2'],
                    'Train_MAPE': metrics_train['MAPE'],
                    'Train_Correlation': metrics_train['Correlation'],
                    'Train_KS_statistic': dist_metrics_train['KS_statistic_mean'],
                    'Train_Distribution_similarity': dist_metrics_train['Distribution_similarity'],
                    
                    # Metriques Test
                    'Test_MAE': metrics_test['MAE'],
                    'Test_RMSE': metrics_test['RMSE'],
                    'Test_MSE': metrics_test['MSE'],
                    'Test_R2': metrics_test['R2'],
                    'Test_MAPE': metrics_test['MAPE'],
                    'Test_Correlation': metrics_test['Correlation'],
                    'Test_KS_statistic': dist_metrics_test['KS_statistic_mean'],
                    'Test_Distribution_similarity': dist_metrics_test['Distribution_similarity'],
                    
                    # Hyperparametres
                    'Hyperparameters': str(method.get_params())
                }
                
                all_results.append(result)
                
                print(f"OK (MAE: {metrics_test['MAE']:.3f}, Time: {method.execution_time:.2f}s)")
                
            except Exception as e:
                print(f"ERREUR: {str(e)}")
                continue
    
    # Compilation resultats
    results_df = pd.DataFrame(all_results)
    
    # Calcul moyennes et ecarts-types
    summary_stats = results_df.groupby(['Mechanism', 'Method']).agg({
        'Test_MAE': ['mean', 'std'],
        'Test_RMSE': ['mean', 'std'],
        'Test_R2': ['mean', 'std'],
        'Test_Correlation': ['mean', 'std'],
        'Test_Distribution_similarity': ['mean', 'std'],
        'Execution_Time': ['mean', 'std']
    }).round(4)
    
    print(f"\n{'=' * 80}")
    print("RESUME DES PERFORMANCES")
    print(f"{'=' * 80}")
    print(summary_stats)
    
    detailed_results[missing_mechanism] = {
        'full_results': results_df,
        'summary': summary_stats,
        'mask': mask_missing
    }
    
    return results_df, detailed_results


# ==============================================================================
# CELLULE 7: VISUALISATIONS
# ==============================================================================

def plot_imputation_results(results_df, save_path='./'):
    """
    Genere toutes les visualisations des resultats
    """
    print(f"\n{'=' * 80}")
    print("GENERATION DES VISUALISATIONS")
    print(f"{'=' * 80}")
    
    mechanisms = results_df['Mechanism'].unique()
    methods = results_df['Method'].unique()
    
    colors = sns.color_palette("husl", len(methods))
    method_colors = dict(zip(methods, colors))
    
    # 1. Comparaison MAE
    fig, axes = plt.subplots(1, len(mechanisms), figsize=(18, 5))
    if len(mechanisms) == 1:
        axes = [axes]
    
    for idx, mech in enumerate(mechanisms):
        data = results_df[results_df['Mechanism'] == mech]
        summary = data.groupby('Method')['Test_MAE'].agg(['mean', 'std']).sort_values('mean')
        
        axes[idx].barh(summary.index, summary['mean'], xerr=summary['std'],
                      color=[method_colors[m] for m in summary.index],
                      alpha=0.8, edgecolor='black')
        axes[idx].set_xlabel('MAE', fontsize=12)
        axes[idx].set_title(f'{mech}', fontsize=14, fontweight='bold')
        axes[idx].grid(axis='x', alpha=0.3)
    
    plt.tight_layout()
    plt.savefig(f'{save_path}comparison_MAE.png', dpi=300, bbox_inches='tight')
    print(f"Graphique sauvegarde: comparison_MAE.png")
    plt.close()
    
    # 2. Comparaison RMSE
    fig, axes = plt.subplots(1, len(mechanisms), figsize=(18, 5))
    if len(mechanisms) == 1:
        axes = [axes]
    
    for idx, mech in enumerate(mechanisms):
        data = results_df[results_df['Mechanism'] == mech]
        summary = data.groupby('Method')['Test_RMSE'].agg(['mean', 'std']).sort_values('mean')
        
        axes[idx].barh(summary.index, summary['mean'], xerr=summary['std'],
                      color=[method_colors[m] for m in summary.index],
                      alpha=0.8, edgecolor='black')
        axes[idx].set_xlabel('RMSE', fontsize=12)
        axes[idx].set_title(f'{mech}', fontsize=14, fontweight='bold')
        axes[idx].grid(axis='x', alpha=0.3)
    
    plt.tight_layout()
    plt.savefig(f'{save_path}comparison_RMSE.png', dpi=300, bbox_inches='tight')
    print(f"Graphique sauvegarde: comparison_RMSE.png")
    plt.close()
    
    # 3. Comparaison R2
    fig, axes = plt.subplots(1, len(mechanisms), figsize=(18, 5))
    if len(mechanisms) == 1:
        axes = [axes]
    
    for idx, mech in enumerate(mechanisms):
        data = results_df[results_df['Mechanism'] == mech]
        summary = data.groupby('Method')['Test_R2'].agg(['mean', 'std']).sort_values('mean', ascending=False)
        
        axes[idx].barh(summary.index, summary['mean'], xerr=summary['std'],
                      color=[method_colors[m] for m in summary.index],
                      alpha=0.8, edgecolor='black')
        axes[idx].set_xlabel('R2', fontsize=12)
        axes[idx].set_title(f'{mech}', fontsize=14, fontweight='bold')
        axes[idx].grid(axis='x', alpha=0.3)
        axes[idx].set_xlim(-0.5, 1.0)
    
    plt.tight_layout()
    plt.savefig(f'{save_path}comparison_R2.png', dpi=300, bbox_inches='tight')
    print(f"Graphique sauvegarde: comparison_R2.png")
    plt.close()
    
    # 4. Temps d'execution
    fig, ax = plt.subplots(figsize=(12, 6))
    
    execution_data = results_df.groupby(['Mechanism', 'Method'])['Execution_Time'].mean().unstack()
    execution_data.plot(kind='bar', ax=ax, color=[method_colors[m] for m in execution_data.columns],
                       edgecolor='black', alpha=0.8)
    
    ax.set_ylabel('Temps d\'execution (secondes)', fontsize=12)
    ax.set_xlabel('Mecanisme de missingness', fontsize=12)
    ax.set_title('Temps d\'execution par methode et mecanisme', fontsize=14, fontweight='bold')
    ax.legend(title='Methode', bbox_to_anchor=(1.05, 1), loc='upper left')
    ax.grid(axis='y', alpha=0.3)
    plt.xticks(rotation=0)
    
    plt.tight_layout()
    plt.savefig(f'{save_path}execution_time.png', dpi=300, bbox_inches='tight')
    print(f"Graphique sauvegarde: execution_time.png")
    plt.close()
    
    # 5. Heatmap des performances
    fig, ax = plt.subplots(figsize=(10, 6))
    
    pivot_data = results_df.groupby(['Method', 'Mechanism'])['Test_MAE'].mean().unstack()
    sns.heatmap(pivot_data, annot=True, fmt='.3f', cmap='RdYlGn_r',
               cbar_kws={'label': 'MAE'}, ax=ax, linewidths=0.5)
    
    ax.set_title('Heatmap des performances MAE', fontsize=14, fontweight='bold')
    ax.set_xlabel('Mecanisme', fontsize=12)
    ax.set_ylabel('Methode', fontsize=12)
    
    plt.tight_layout()
    plt.savefig(f'{save_path}heatmap_MAE.png', dpi=300, bbox_inches='tight')
    print(f"Graphique sauvegarde: heatmap_MAE.png")
    plt.close()
    
    # 6. Distribution similarity
    fig, ax = plt.subplots(figsize=(12, 6))
    
    dist_data = results_df.groupby(['Mechanism', 'Method'])['Test_Distribution_similarity'].mean().unstack()
    dist_data.plot(kind='bar', ax=ax, color=[method_colors[m] for m in dist_data.columns],
                  edgecolor='black', alpha=0.8)
    
    ax.set_ylabel('Similarite de distribution', fontsize=12)
    ax.set_xlabel('Mecanisme', fontsize=12)
    ax.set_title('Similarite des distributions (plus proche de 1 = meilleur)',
                fontsize=14, fontweight='bold')
    ax.legend(title='Methode', bbox_to_anchor=(1.05, 1), loc='upper left')
    ax.grid(axis='y', alpha=0.3)
    ax.set_ylim(0, 1)
    plt.xticks(rotation=0)
    
    plt.tight_layout()
    plt.savefig(f'{save_path}distribution_similarity.png', dpi=300, bbox_inches='tight')
    print(f"Graphique sauvegarde: distribution_similarity.png")
    plt.close()
    
    print(f"\nToutes les visualisations generees avec succes!")


def plot_missingness_patterns(df_missing, mask, save_path='./'):
    """
    Visualise les patterns de donnees manquantes
    """
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))
    
    # Heatmap des donnees manquantes (sample)
    mask_sample = mask.iloc[:100]  # Premier 100 observations
    sns.heatmap(mask_sample.T, cbar=False, cmap='RdYlGn_r', ax=axes[0], yticklabels=True)
    axes[0].set_title('Pattern des donnees manquantes (100 premiers patients)', fontsize=14, fontweight='bold')
    axes[0].set_xlabel('Observations', fontsize=12)
    axes[0].set_ylabel('Variables', fontsize=12)
    
    # Taux de missingness par variable
    missing_rates = (mask.sum() / len(mask)) * 100
    missing_rates = missing_rates.sort_values(ascending=True)
    
    axes[1].barh(range(len(missing_rates)), missing_rates.values,
                color='coral', edgecolor='black', alpha=0.8)
    axes[1].set_yticks(range(len(missing_rates)))
    axes[1].set_yticklabels(missing_rates.index, fontsize=8)
    axes[1].set_xlabel('Taux de donnees manquantes (%)', fontsize=12)
    axes[1].set_title('Taux de missingness par variable', fontsize=14, fontweight='bold')
    axes[1].grid(axis='x', alpha=0.3)
    
    plt.tight_layout()
    plt.savefig(f'{save_path}missingness_patterns.png', dpi=300, bbox_inches='tight')
    print(f"Graphique sauvegarde: missingness_patterns.png")
    plt.close()


# ==============================================================================
# CELLULE 8: ANALYSE DE SENSIBILITE
# ==============================================================================

def sensitivity_analysis(df_complete, missing_rates=[0.10, 0.20, 0.30, 0.40],
                        mechanisms=['MCAR', 'MAR', 'MNAR'], n_iterations=3, seed=42):
    """
    Analyse de sensibilite: evalue la performance en fonction du taux de missingness
    """
    print(f"\n{'=' * 80}")
    print("ANALYSE DE SENSIBILITE")
    print(f"{'=' * 80}")
    print(f"Configurations a tester: {len(missing_rates)} taux x {len(mechanisms)} mecanismes")
    print(f"Iterations par configuration: {n_iterations}")
    
    all_sensitivity_results = []
    
    for mech in mechanisms:
        for rate in missing_rates:
            print(f"\n{'-' * 80}")
            print(f"Configuration: {mech} - {rate:.0%} missingness")
            print(f"{'-' * 80}")
            
            # Evaluation
            results_df, _ = evaluate_imputation_pipeline(
                df_complete,
                missing_mechanism=mech,
                missing_rate=rate,
                n_iterations=n_iterations,
                seed=seed
            )
            
            results_df['Missing_Rate'] = rate
            all_sensitivity_results.append(results_df)
    
    # Compilation
    sensitivity_df = pd.concat(all_sensitivity_results, ignore_index=True)
    
    # Visualisation
    print(f"\n{'=' * 80}")
    print("VISUALISATION ANALYSE DE SENSIBILITE")
    print(f"{'=' * 80}")
    
    for mech in mechanisms:
        fig, axes = plt.subplots(1, 3, figsize=(18, 5))
        
        data_mech = sensitivity_df[sensitivity_df['Mechanism'] == mech]
        
        # MAE vs Missing Rate
        for method in data_mech['Method'].unique():
            method_data = data_mech[data_mech['Method'] == method]
            summary = method_data.groupby('Missing_Rate')['Test_MAE'].agg(['mean', 'std'])
            
            axes[0].plot(summary.index * 100, summary['mean'], marker='o', label=method, linewidth=2)
            axes[0].fill_between(summary.index * 100,
                                summary['mean'] - summary['std'],
                                summary['mean'] + summary['std'],
                                alpha=0.2)
        
        axes[0].set_xlabel('Taux de donnees manquantes (%)', fontsize=12)
        axes[0].set_ylabel('MAE', fontsize=12)
        axes[0].set_title(f'{mech} - MAE', fontsize=14, fontweight='bold')
        axes[0].legend(fontsize=8)
        axes[0].grid(alpha=0.3)
        
        # RMSE vs Missing Rate
        for method in data_mech['Method'].unique():
            method_data = data_mech[data_mech['Method'] == method]
            summary = method_data.groupby('Missing_Rate')['Test_RMSE'].agg(['mean', 'std'])
            
            axes[1].plot(summary.index * 100, summary['mean'], marker='o', label=method, linewidth=2)
            axes[1].fill_between(summary.index * 100,
                                summary['mean'] - summary['std'],
                                summary['mean'] + summary['std'],
                                alpha=0.2)
        
        axes[1].set_xlabel('Taux de donnees manquantes (%)', fontsize=12)
        axes[1].set_ylabel('RMSE', fontsize=12)
        axes[1].set_title(f'{mech} - RMSE', fontsize=14, fontweight='bold')
        axes[1].legend(fontsize=8)
        axes[1].grid(alpha=0.3)
        
        # R2 vs Missing Rate
        for method in data_mech['Method'].unique():
            method_data = data_mech[data_mech['Method'] == method]
            summary = method_data.groupby('Missing_Rate')['Test_R2'].agg(['mean', 'std'])
            
            axes[2].plot(summary.index * 100, summary['mean'], marker='o', label=method, linewidth=2)
            axes[2].fill_between(summary.index * 100,
                                summary['mean'] - summary['std'],
                                summary['mean'] + summary['std'],
                                alpha=0.2)
        
        axes[2].set_xlabel('Taux de donnees manquantes (%)', fontsize=12)
        axes[2].set_ylabel('R2', fontsize=12)
        axes[2].set_title(f'{mech} - R2', fontsize=14, fontweight='bold')
        axes[2].legend(fontsize=8)
        axes[2].grid(alpha=0.3)
        
        plt.tight_layout()
        plt.savefig(f'sensitivity_analysis_{mech}.png', dpi=300, bbox_inches='tight')
        print(f"Graphique sauvegarde: sensitivity_analysis_{mech}.png")
        plt.close()
    
    return sensitivity_df


# ==============================================================================
# CELLULE 9: RAPPORT FINAL ET SAUVEGARDE
# ==============================================================================

def generate_final_report(all_results, sensitivity_results, save_path='./'):
    """
    Genere un rapport final complet avec sauvegarde
    """
    print(f"\n{'=' * 80}")
    print("GENERATION DU RAPPORT FINAL")
    print(f"{'=' * 80}")
    
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    
    # Sauvegarde des resultats complets
    all_results.to_csv(f'{save_path}imputation_results_{timestamp}.csv', index=False)
    print(f"Resultats sauvegardes: imputation_results_{timestamp}.csv")
    
    sensitivity_results.to_csv(f'{save_path}sensitivity_results_{timestamp}.csv', index=False)
    print(f"Sensitivite sauvegardee: sensitivity_results_{timestamp}.csv")
    
    # Generation du rapport textuel
    report_lines = []
    report_lines.append("=" * 80)
    report_lines.append("RAPPORT D'ANALYSE D'IMPUTATION")
    report_lines.append("Evaluations Cognitives Longitudinales - Maladie de Parkinson")
    report_lines.append("=" * 80)
    report_lines.append(f"\nDate: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
    report_lines.append(f"\nNombre total d'observations: {len(all_results)}")
    report_lines.append(f"Mecanismes testes: {', '.join(all_results['Mechanism'].unique())}")
    report_lines.append(f"Methodes evaluees: {', '.join(all_results['Method'].unique())}")
    
    # Resume par mecanisme
    for mech in all_results['Mechanism'].unique():
        report_lines.append(f"\n\n{'-' * 80}")
        report_lines.append(f"RESULTATS: {mech}")
        report_lines.append(f"{'-' * 80}")
        
        data_mech = all_results[all_results['Mechanism'] == mech]
        summary = data_mech.groupby('Method').agg({
            'Test_MAE': ['mean', 'std'],
            'Test_RMSE': ['mean', 'std'],
            'Test_R2': ['mean', 'std'],
            'Execution_Time': 'mean'
        }).round(4)
        
        report_lines.append("\n" + summary.to_string())
        
        # Meilleure methode
        best_method_mae = data_mech.groupby('Method')['Test_MAE'].mean().idxmin()
        best_method_r2 = data_mech.groupby('Method')['Test_R2'].mean().idxmax()
        
        report_lines.append(f"\n\nMeilleure methode (MAE): {best_method_mae}")
        report_lines.append(f"Meilleure methode (R2): {best_method_r2}")
    
    # Recommandations
    report_lines.append(f"\n\n{'=' * 80}")
    report_lines.append("RECOMMANDATIONS")
    report_lines.append(f"{'=' * 80}")
    
    overall_best = all_results.groupby('Method')['Test_MAE'].mean().idxmin()
    overall_best_mae = all_results.groupby('Method')['Test_MAE'].mean().min()
    
    report_lines.append(f"\nMethode globalement recommandee: {overall_best}")
    report_lines.append(f"  - MAE moyen: {overall_best_mae:.4f}")
    report_lines.append(f"\nCette methode offre le meilleur compromis entre precision et robustesse")
    report_lines.append(f"a travers les differents mecanismes de donnees manquantes.")
    
    # Considerations pratiques
    report_lines.append(f"\n\nCONSIDERATIONS PRATIQUES:")
    
    for method in all_results['Method'].unique():
        method_data = all_results[all_results['Method'] == method]
        avg_time = method_data['Execution_Time'].mean()
        avg_mae = method_data['Test_MAE'].mean()
        
        report_lines.append(f"\n  {method}:")
        report_lines.append(f"    - Temps d'execution moyen: {avg_time:.2f}s")
        report_lines.append(f"    - MAE moyen: {avg_mae:.4f}")
    
    # Sauvegarde du rapport
    report_text = "\n".join(report_lines)
    
    with open(f'{save_path}rapport_final_{timestamp}.txt', 'w', encoding='utf-8') as f:
        f.write(report_text)
    
    print(f"Rapport textuel sauvegarde: rapport_final_{timestamp}.txt")
    
    # Sauvegarde JSON pour analyses ulterieures
    results_dict = {
        'timestamp': timestamp,
        'mechanisms': all_results['Mechanism'].unique().tolist(),
        'methods': all_results['Method'].unique().tolist(),
        'summary': all_results.groupby('Method')['Test_MAE'].agg(['mean', 'std']).to_dict(),
        'best_method': overall_best,
        'best_mae': float(overall_best_mae)
    }
    
    with open(f'{save_path}results_summary_{timestamp}.json', 'w') as f:
        json.dump(results_dict, f, indent=2)
    
    print(f"Resumé JSON sauvegarde: results_summary_{timestamp}.json")
    
    # Affichage du rapport
    print("\n" + report_text)
    
    return report_text


# ==============================================================================
# CELLULE 10: EXECUTION PRINCIPALE
# ==============================================================================

def main():
    """
    Fonction principale d'execution
    """
    print("\n")
    print("=" * 80)
    print("=" * 80)
    print("     ANALYSE COMPLETE D'IMPUTATION - DONNEES NEUROPSYCHOLOGIQUES")
    print("=" * 80)
    print("=" * 80)
    
    # Configuration
    DATA_PATH = "Without_Missing_Data.xlsx"
    MISSING_RATE = 0.20
    N_ITERATIONS = 5
    TEST_SIZE = 0.20
    SEED = 42
    
    # ETAPE 1: Chargement des donnees
    print("\n[ETAPE 1/5] Chargement et preparation des donnees...")
    df_numeric, df_original, metadata = load_and_prepare_data(DATA_PATH)
    
    # ETAPE 2: Evaluations principales (MAR, MCAR, MNAR)
    print("\n[ETAPE 2/5] Evaluations des methodes d'imputation...")
    
    all_results = []
    all_detailed_results = {}
    
    for mechanism in ['MCAR', 'MAR', 'MNAR']:
        results_df, detailed = evaluate_imputation_pipeline(
            df_numeric,
            missing_mechanism=mechanism,
            missing_rate=MISSING_RATE,
            n_iterations=N_ITERATIONS,
            seed=SEED
        )
        all_results.append(results_df)
        all_detailed_results[mechanism] = detailed
    
    # Compilation resultats
    compiled_results = pd.concat(all_results, ignore_index=True)
    
    # ETAPE 3: Visualisations
    print("\n[ETAPE 3/5] Generation des visualisations...")
    plot_imputation_results(compiled_results, save_path='./')
    
    # Visualisation des patterns de missingness
    df_missing_example, mask_example, _ = simulate_missing_data(
        df_numeric, missing_rate=MISSING_RATE, mechanism='MAR', seed=SEED
    )
    plot_missingness_patterns(df_missing_example, mask_example, save_path='./')
    
    # ETAPE 4: Analyse de sensibilite
    print("\n[ETAPE 4/5] Analyse de sensibilite...")
    sensitivity_results = sensitivity_analysis(
        df_numeric,
        missing_rates=[0.10, 0.20, 0.30, 0.40],
        mechanisms=['MCAR', 'MAR', 'MNAR'],
        n_iterations=3,
        seed=SEED
    )
    
    # ETAPE 5: Rapport final
    print("\n[ETAPE 5/5] Generation du rapport final...")
    final_report = generate_final_report(compiled_results, sensitivity_results, save_path='./')
    
    print("\n" + "=" * 80)
    print("ANALYSE TERMINEE AVEC SUCCES!")
    print("=" * 80)
    print("\nFichiers generes:")
    print("  - imputation_results_YYYYMMDD_HHMMSS.csv")
    print("  - sensitivity_results_YYYYMMDD_HHMMSS.csv")
    print("  - rapport_final_YYYYMMDD_HHMMSS.txt")
    print("  - results_summary_YYYYMMDD_HHMMSS.json")
    print("  - comparison_MAE.png")
    print("  - comparison_RMSE.png")
    print("  - comparison_R2.png")
    print("  - execution_time.png")
    print("  - heatmap_MAE.png")
    print("  - distribution_similarity.png")
    print("  - missingness_patterns.png")
    print("  - sensitivity_analysis_MCAR.png")
    print("  - sensitivity_analysis_MAR.png")
    print("  - sensitivity_analysis_MNAR.png")
    print("=" * 80)
    
    return compiled_results, sensitivity_results, all_detailed_results


# ==============================================================================
# EXECUTION
# ==============================================================================

if __name__ == "__main__":
    # Montage Google Drive (si necessaire)
    try:
        from google.colab import drive
        drive.mount('/content/drive')
    except:
        pass
    
    # Execution du pipeline complet
    results, sensitivity, detailed = main()
    
    print("\nPipeline d'analyse termine!")
    print("\nPour acceder aux resultats:")
    print("  - results: DataFrame avec tous les resultats d'imputation")
    print("  - sensitivity: DataFrame avec analyse de sensibilite")
    print("  - detailed: Dictionnaire avec resultats detailles par mecanisme")
```

```python
# ==============================================================================
# CODE DE VISUALISATION AVEC DETECTION AUTOMATIQUE DES FICHIERS
# ==============================================================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import os
from pathlib import Path

# Configuration
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print("=" * 80)
print("GENERATION DES VISUALISATIONS - VERSION AUTOMATIQUE")
print("=" * 80)

# ==============================================================================
# ETAPE 0: DETECTION AUTOMATIQUE DES FICHIERS
# ==============================================================================

def find_results_files(directory="."):
    """
    Trouve automatiquement les fichiers de résultats
    """
    print("\nRecherche des fichiers de resultats...")
    
    # Changer vers le répertoire spécifié
    os.chdir(directory)
    
    # Chercher tous les CSV
    all_csv = glob.glob("*.csv")
    print(f"\nFichiers CSV disponibles ({len(all_csv)}):")
    for f in all_csv[:10]:  # Afficher les 10 premiers
        print(f"  - {f}")
    if len(all_csv) > 10:
        print(f"  ... et {len(all_csv) - 10} autres")
    
    # Chercher fichiers de résultats
    results_files = sorted(glob.glob("*imputation*results*.csv"))
    results_files = [f for f in results_files if 'sensitivity' not in f.lower()]
    
    if results_files:
        print(f"\nFichiers de resultats principaux trouves:")
        for f in results_files:
            size = os.path.getsize(f) / 1024  # Ko
            print(f"  - {f} ({size:.1f} Ko)")
        results_path = results_files[-1]  # Le plus récent
        print(f"\n>>> Selection: {results_path}")
    else:
        print("\nERREUR: Aucun fichier de resultats trouve!")
        print("Le fichier doit contenir 'imputation' et 'results' dans son nom")
        return None, None
    
    # Chercher fichiers de sensibilité
    sensitivity_files = sorted(glob.glob("*sensitivity*.csv"))
    
    if sensitivity_files:
        print(f"\nFichiers de sensibilite trouves:")
        for f in sensitivity_files:
            size = os.path.getsize(f) / 1024
            print(f"  - {f} ({size:.1f} Ko)")
        sensitivity_path = sensitivity_files[-1]
        print(f"\n>>> Selection: {sensitivity_path}")
    else:
        print("\nAucun fichier de sensibilite trouve (optionnel)")
        sensitivity_path = None
    
    return results_path, sensitivity_path


def load_results(results_path, sensitivity_path=None):
    """
    Charge les résultats depuis les fichiers CSV
    """
    print("\n" + "=" * 80)
    print("CHARGEMENT DES DONNEES")
    print("=" * 80)
    
    # Chargement résultats principaux
    try:
        results_df = pd.read_csv(results_path)
        print(f"\nOK - Resultats principaux charges: {len(results_df)} lignes")
        print(f"     Colonnes: {list(results_df.columns[:5])}...")
        print(f"     Mecanismes: {results_df['Mechanism'].unique()}")
        print(f"     Methodes: {results_df['Method'].unique()}")
    except Exception as e:
        print(f"\nERREUR lors du chargement de {results_path}:")
        print(f"  {str(e)}")
        return None, None
    
    # Chargement sensibilité
    if sensitivity_path and Path(sensitivity_path).exists():
        try:
            sensitivity_df = pd.read_csv(sensitivity_path)
            print(f"\nOK - Analyse de sensibilite chargee: {len(sensitivity_df)} lignes")
        except Exception as e:
            print(f"\nAvertissement: Erreur lors du chargement de la sensibilite:")
            print(f"  {str(e)}")
            sensitivity_df = None
    else:
        sensitivity_df = None
        print("\nAnalyse de sensibilite non disponible")
    
    return results_df, sensitivity_df


# ==============================================================================
# FONCTIONS DE VISUALISATION (identiques au code précédent)
# ==============================================================================

def plot_comparison_mae(results_df, save_path='./'):
    """Comparaison MAE"""
    print("\n  Generation: Comparaison MAE...")
    
    mechanisms = results_df['Mechanism'].unique()
    methods = results_df['Method'].unique()
    
    colors = sns.color_palette("husl", len(methods))
    method_colors = dict(zip(methods, colors))
    
    fig, axes = plt.subplots(1, len(mechanisms), figsize=(18, 5))
    if len(mechanisms) == 1:
        axes = [axes]
    
    for idx, mech in enumerate(mechanisms):
        data = results_df[results_df['Mechanism'] == mech]
        summary = data.groupby('Method')['Test_MAE'].agg(['mean', 'std']).sort_values('mean')
        
        axes[idx].barh(summary.index, summary['mean'], xerr=summary['std'],
                      color=[method_colors.get(m, 'gray') for m in summary.index],
                      alpha=0.8, edgecolor='black')
        axes[idx].set_xlabel('MAE', fontsize=12)
        axes[idx].set_title(f'{mech}', fontsize=14, fontweight='bold')
        axes[idx].grid(axis='x', alpha=0.3)
    
    plt.tight_layout()
    plt.savefig(f'{save_path}comparison_MAE.png', dpi=300, bbox_inches='tight')
    print(f"    OK - Sauvegarde: comparison_MAE.png")
    plt.close()


def plot_comparison_rmse(results_df, save_path='./'):
    """Comparaison RMSE"""
    print("\n  Generation: Comparaison RMSE...")
    
    mechanisms = results_df['Mechanism'].unique()
    methods = results_df['Method'].unique()
    
    colors = sns.color_palette("husl", len(methods))
    method_colors = dict(zip(methods, colors))
    
    fig, axes = plt.subplots(1, len(mechanisms), figsize=(18, 5))
    if len(mechanisms) == 1:
        axes = [axes]
    
    for idx, mech in enumerate(mechanisms):
        data = results_df[results_df['Mechanism'] == mech]
        summary = data.groupby('Method')['Test_RMSE'].agg(['mean', 'std']).sort_values('mean')
        
        axes[idx].barh(summary.index, summary['mean'], xerr=summary['std'],
                      color=[method_colors.get(m, 'gray') for m in summary.index],
                      alpha=0.8, edgecolor='black')
        axes[idx].set_xlabel('RMSE', fontsize=12)
        axes[idx].set_title(f'{mech}', fontsize=14, fontweight='bold')
        axes[idx].grid(axis='x', alpha=0.3)
    
    plt.tight_layout()
    plt.savefig(f'{save_path}comparison_RMSE.png', dpi=300, bbox_inches='tight')
    print(f"    OK - Sauvegarde: comparison_RMSE.png")
    plt.close()


def plot_comparison_r2(results_df, save_path='./'):
    """Comparaison R²"""
    print("\n  Generation: Comparaison R2...")
    
    mechanisms = results_df['Mechanism'].unique()
    methods = results_df['Method'].unique()
    
    colors = sns.color_palette("husl", len(methods))
    method_colors = dict(zip(methods, colors))
    
    fig, axes = plt.subplots(1, len(mechanisms), figsize=(18, 5))
    if len(mechanisms) == 1:
        axes = [axes]
    
    for idx, mech in enumerate(mechanisms):
        data = results_df[results_df['Mechanism'] == mech]
        summary = data.groupby('Method')['Test_R2'].agg(['mean', 'std']).sort_values('mean', ascending=False)
        
        axes[idx].barh(summary.index, summary['mean'], xerr=summary['std'],
                      color=[method_colors.get(m, 'gray') for m in summary.index],
                      alpha=0.8, edgecolor='black')
        axes[idx].set_xlabel('R²', fontsize=12)
        axes[idx].set_title(f'{mech}', fontsize=14, fontweight='bold')
        axes[idx].grid(axis='x', alpha=0.3)
        axes[idx].set_xlim(-0.5, 1.0)
    
    plt.tight_layout()
    plt.savefig(f'{save_path}comparison_R2.png', dpi=300, bbox_inches='tight')
    print(f"    OK - Sauvegarde: comparison_R2.png")
    plt.close()


def plot_execution_time(results_df, save_path='./'):
    """Temps d'exécution"""
    print("\n  Generation: Temps d'execution...")
    
    methods = results_df['Method'].unique()
    colors = sns.color_palette("husl", len(methods))
    method_colors = dict(zip(methods, colors))
    
    fig, ax = plt.subplots(figsize=(12, 6))
    
    execution_data = results_df.groupby(['Mechanism', 'Method'])['Execution_Time'].mean().unstack()
    execution_data.plot(kind='bar', ax=ax,
                       color=[method_colors.get(m, 'gray') for m in execution_data.columns],
                       edgecolor='black', alpha=0.8)
    
    ax.set_ylabel('Temps d\'execution (secondes)', fontsize=12)
    ax.set_xlabel('Mecanisme de missingness', fontsize=12)
    ax.set_title('Temps d\'execution par methode et mecanisme', fontsize=14, fontweight='bold')
    ax.legend(title='Methode', bbox_to_anchor=(1.05, 1), loc='upper left')
    ax.grid(axis='y', alpha=0.3)
    plt.xticks(rotation=0)
    
    plt.tight_layout()
    plt.savefig(f'{save_path}execution_time.png', dpi=300, bbox_inches='tight')
    print(f"    OK - Sauvegarde: execution_time.png")
    plt.close()


def plot_heatmap_mae(results_df, save_path='./'):
    """Heatmap MAE"""
    print("\n  Generation: Heatmap MAE...")
    
    fig, ax = plt.subplots(figsize=(10, 6))
    
    pivot_data = results_df.groupby(['Method', 'Mechanism'])['Test_MAE'].mean().unstack()
    sns.heatmap(pivot_data, annot=True, fmt='.3f', cmap='RdYlGn_r',
               cbar_kws={'label': 'MAE'}, ax=ax, linewidths=0.5)
    
    ax.set_title('Heatmap des performances MAE', fontsize=14, fontweight='bold')
    ax.set_xlabel('Mecanisme', fontsize=12)
    ax.set_ylabel('Methode', fontsize=12)
    
    plt.tight_layout()
    plt.savefig(f'{save_path}heatmap_MAE.png', dpi=300, bbox_inches='tight')
    print(f"    OK - Sauvegarde: heatmap_MAE.png")
    plt.close()


def plot_sensitivity_analysis(sensitivity_df, save_path='./'):
    """Analyse de sensibilité"""
    if sensitivity_df is None or sensitivity_df.empty:
        print("\n  Analyse de sensibilite non disponible - skip")
        return
    
    print("\n  Generation: Analyse de sensibilite...")
    
    mechanisms = sensitivity_df['Mechanism'].unique()
    
    for mech in mechanisms:
        print(f"    Mecanisme: {mech}...")
        
        fig, axes = plt.subplots(1, 3, figsize=(18, 5))
        data_mech = sensitivity_df[sensitivity_df['Mechanism'] == mech]
        
        # MAE
        for method in data_mech['Method'].unique():
            method_data = data_mech[data_mech['Method'] == method]
            summary = method_data.groupby('Missing_Rate')['Test_MAE'].agg(['mean', 'std'])
            
            x_vals = (summary.index * 100).values
            y_mean = summary['mean'].values
            y_std = summary['std'].fillna(0).values
            
            axes[0].plot(x_vals, y_mean, marker='o', label=method, linewidth=2)
            axes[0].fill_between(x_vals, y_mean - y_std, y_mean + y_std, alpha=0.2)
        
        axes[0].set_xlabel('Taux de donnees manquantes (%)', fontsize=12)
        axes[0].set_ylabel('MAE', fontsize=12)
        axes[0].set_title(f'{mech} - MAE', fontsize=14, fontweight='bold')
        axes[0].legend(fontsize=8)
        axes[0].grid(alpha=0.3)
        
        # RMSE
        for method in data_mech['Method'].unique():
            method_data = data_mech[data_mech['Method'] == method]
            summary = method_data.groupby('Missing_Rate')['Test_RMSE'].agg(['mean', 'std'])
            
            x_vals = (summary.index * 100).values
            y_mean = summary['mean'].values
            y_std = summary['std'].fillna(0).values
            
            axes[1].plot(x_vals, y_mean, marker='o', label=method, linewidth=2)
            axes[1].fill_between(x_vals, y_mean - y_std, y_mean + y_std, alpha=0.2)
        
        axes[1].set_xlabel('Taux de donnees manquantes (%)', fontsize=12)
        axes[1].set_ylabel('RMSE', fontsize=12)
        axes[1].set_title(f'{mech} - RMSE', fontsize=14, fontweight='bold')
        axes[1].legend(fontsize=8)
        axes[1].grid(alpha=0.3)
        
        # R²
        for method in data_mech['Method'].unique():
            method_data = data_mech[data_mech['Method'] == method]
            summary = method_data.groupby('Missing_Rate')['Test_R2'].agg(['mean', 'std'])
            
            x_vals = (summary.index * 100).values
            y_mean = summary['mean'].values
            y_std = summary['std'].fillna(0).values
            
            axes[2].plot(x_vals, y_mean, marker='o', label=method, linewidth=2)
            axes[2].fill_between(x_vals, y_mean - y_std, y_mean + y_std, alpha=0.2)
        
        axes[2].set_xlabel('Taux de donnees manquantes (%)', fontsize=12)
        axes[2].set_ylabel('R²', fontsize=12)
        axes[2].set_title(f'{mech} - R²', fontsize=14, fontweight='bold')
        axes[2].legend(fontsize=8)
        axes[2].grid(alpha=0.3)
        
        plt.tight_layout()
        plt.savefig(f'{save_path}sensitivity_analysis_{mech}.png', dpi=300, bbox_inches='tight')
        print(f"      OK - Sauvegarde: sensitivity_analysis_{mech}.png")
        plt.close()


def generate_summary_tables(results_df, save_path='./'):
    """Tableaux récapitulatifs"""
    print("\n  Generation: Tableaux recapitulatifs...")
    
    # Résumé par mécanisme
    summary = results_df.groupby(['Mechanism', 'Method']).agg({
        'Test_MAE': ['mean', 'std'],
        'Test_RMSE': ['mean', 'std'],
        'Test_R2': ['mean', 'std'],
        'Execution_Time': ['mean', 'std']
    }).round(4)
    
    summary.to_csv(f'{save_path}summary_by_mechanism.csv')
    print(f"    OK - Sauvegarde: summary_by_mechanism.csv")
    
    # Meilleure méthode par mécanisme
    best_methods = []
    for mech in results_df['Mechanism'].unique():
        data = results_df[results_df['Mechanism'] == mech]
        best_mae = data.groupby('Method')['Test_MAE'].mean().idxmin()
        best_r2 = data.groupby('Method')['Test_R2'].mean().idxmax()
        
        best_methods.append({
            'Mechanism': mech,
            'Best_MAE': best_mae,
            'Best_R2': best_r2
        })
    
    best_df = pd.DataFrame(best_methods)
    best_df.to_csv(f'{save_path}best_methods_by_criterion.csv', index=False)
    print(f"    OK - Sauvegarde: best_methods_by_criterion.csv")
    
    return summary, best_df


# ==============================================================================
# FONCTION PRINCIPALE
# ==============================================================================

def main():
    """
    Fonction principale avec détection automatique
    """
    print("\n" + "=" * 80)
    print("GENERATION COMPLETE DES VISUALISATIONS")
    print("=" * 80)
    
    # CONFIGURATION - MODIFIER SI NECESSAIRE
    DIRECTORY = "."  # Répertoire où chercher les fichiers
    SAVE_PATH = "./"  # Où sauvegarder les graphiques
    
    # Détection automatique des fichiers
    results_path, sensitivity_path = find_results_files(DIRECTORY)
    
    if results_path is None:
        print("\n" + "=" * 80)
        print("ERREUR: Impossible de trouver les fichiers de resultats")
        print("=" * 80)
        print("\nVerifiez que les fichiers CSV sont bien dans le repertoire:")
        print(f"  {os.getcwd()}")
        print("\nLe fichier doit contenir 'imputation' et 'results' dans son nom")
        return
    
    # Chargement
    results_df, sensitivity_df = load_results(results_path, sensitivity_path)
    
    if results_df is None:
        print("\nERREUR: Impossible de charger les resultats")
        return
    
    # Visualisations
    print("\n" + "=" * 80)
    print("GENERATION DES GRAPHIQUES")
    print("=" * 80)
    
    try:
        plot_comparison_mae(results_df, SAVE_PATH)
        plot_comparison_rmse(results_df, SAVE_PATH)
        plot_comparison_r2(results_df, SAVE_PATH)
        plot_execution_time(results_df, SAVE_PATH)
        plot_heatmap_mae(results_df, SAVE_PATH)
        
        # Tableaux
        print("\n" + "=" * 80)
        print("GENERATION DES TABLEAUX")
        print("=" * 80)
        summary, best = generate_summary_tables(results_df, SAVE_PATH)
        
        # Sensibilité
        if sensitivity_df is not None:
            print("\n" + "=" * 80)
            print("ANALYSE DE SENSIBILITE")
            print("=" * 80)
            plot_sensitivity_analysis(sensitivity_df, SAVE_PATH)
        
        print("\n" + "=" * 80)
        print("GENERATION TERMINEE AVEC SUCCES")
        print("=" * 80)
        
        print("\nFichiers generes dans:", os.getcwd())
        print("  - comparison_MAE.png")
        print("  - comparison_RMSE.png")
        print("  - comparison_R2.png")
        print("  - execution_time.png")
        print("  - heatmap_MAE.png")
        print("  - summary_by_mechanism.csv")
        print("  - best_methods_by_criterion.csv")
        
        if sensitivity_df is not None:
            for mech in sensitivity_df['Mechanism'].unique():
                print(f"  - sensitivity_analysis_{mech}.png")
        
    except Exception as e:
        print(f"\nERREUR lors de la generation des graphiques:")
        print(f"  {str(e)}")
        import traceback
        traceback.print_exc()


if __name__ == "__main__":
    main()
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
================================================================================
IMPUTATION PRODUCTION - MISSFOREST OPTIMISE
================================================================================
Applique MissForest avec hyperparametres optimaux sur donnees reelles
Fichier: With_Missing_Data.xlsx
CORRECTION FINALE : Gestion des colonnes et debug complet
================================================================================
"""

import os
import warnings
import numpy as np
import pandas as pd
from datetime import datetime
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import RandomForestRegressor

warnings.filterwarnings('ignore')

print("=" * 80)
print("IMPUTATION PRODUCTION - MISSFOREST")
print("=" * 80)


# ==============================================================================
# CONFIGURATION OPTIMALE (d'apres resultats)
# ==============================================================================
HYPERPARAMETERS = {
    'n_estimators': 100, #50,
    'max_iter': 5, #3,
    'random_state': 42,
    'n_jobs': -1,
    'verbose': 1
}

MAE_EXPECTED = 2.20


# ==============================================================================
# CHARGEMENT DES DONNEES
# ==============================================================================
def load_data_with_missing(filepath):
    """Charge les donnees avec missing - CORRIGE"""
    print(f"\n[1/5] Chargement: {filepath}")
    
    df = pd.read_excel(filepath, sheet_name='NS-Park')
    print(f"  Dataset initial: {df.shape[0]} obs, {df.shape[1]} vars")
    
    # Colonnes ID (non a imputer)
    id_cols = ['Identifiant', 'date exam', 'date naiss']
    
    # Identifier colonnes datetime et les exclure
    datetime_cols = df.select_dtypes(include=['datetime64']).columns.tolist()
    print(f"  Colonnes datetime a exclure: {datetime_cols}")
    
    # Colonnes a conserver (seulement numeriques + Age, Sexe, dur.Sco)
    demographic_cols = ['Age', 'Sexe', 'dur.Sco']
    
    # Variables cognitives (toutes sauf ID et demographics)
    cognitive_cols = [
        c for c in df.columns 
        if c not in id_cols + demographic_cols and c not in datetime_cols
    ]
    
    print(f"  Nombre de variables cognitives: {len(cognitive_cols)}")
    
    # Preparation
    df_work = df.copy()
    
    # Conversion numerique
    df_work['Age'] = pd.to_numeric(df_work['Age'], errors='coerce')
    df_work['dur.Sco'] = pd.to_numeric(df_work['dur.Sco'], errors='coerce')
    
    # Encodage Sexe (Homme=1, Femme=0)
    df_work['Sexe_binary'] = (df_work['Sexe'] == 'Homme').astype(int)
    
    # Selection colonnes a imputer
    numeric_cols = ['Age', 'dur.Sco', 'Sexe_binary'] + cognitive_cols
    
    # Extraction et conversion
    df_numeric = df_work[numeric_cols].copy()
    
    # Conversion forcee en numeric (elimine tout type residuel)
    for col in df_numeric.columns:
        df_numeric[col] = pd.to_numeric(df_numeric[col], errors='coerce')
    
    # CORRECTION : Identifier et supprimer colonnes problematiques
    print(f"\n  Analyse des colonnes:")
    problematic_cols = []
    for col in df_numeric.columns:
        non_null_count = df_numeric[col].notna().sum()
        non_null_pct = non_null_count / len(df_numeric) * 100
        unique_vals = df_numeric[col].nunique()
        
        # Colonnes entierement nulles ou constantes
        if non_null_count == 0:
            print(f"    - {col}: ENTIEREMENT NULL (exclusion)")
            problematic_cols.append(col)
        elif unique_vals == 1:
            print(f"    - {col}: CONSTANTE (exclusion)")
            problematic_cols.append(col)
        elif non_null_pct < 5:
            print(f"    - {col}: {non_null_pct:.1f}% non-null (exclusion)")
            problematic_cols.append(col)
    
    # Supprimer colonnes problematiques
    if problematic_cols:
        df_numeric = df_numeric.drop(columns=problematic_cols)
        print(f"\n  Colonnes problematiques supprimees: {len(problematic_cols)}")
    
    print(f"  Dataset final pour imputation: {df_numeric.shape[0]} obs, {df_numeric.shape[1]} vars")
    print(f"  Types de colonnes:")
    print(f"    {df_numeric.dtypes.value_counts()}")
    
    # Analyse missing
    n_missing = df_numeric.isnull().sum().sum()
    missing_rate = n_missing / (df_numeric.shape[0] * df_numeric.shape[1])
    
    print(f"\n  Valeurs manquantes: {n_missing:,} ({missing_rate*100:.2f}%)")
    print(f"  Top 10 variables avec missing:")
    
    missing_per_col = df_numeric.isnull().sum()
    missing_per_col = missing_per_col[missing_per_col > 0].sort_values(ascending=False)
    for col, n in missing_per_col.head(10).items():
        print(f"    - {col:25s}: {n:4d} ({n/len(df_numeric)*100:5.1f}%)")
    
    metadata = {
        'n_obs': len(df_numeric),
        'n_vars': len(df_numeric.columns),
        'n_missing': n_missing,
        'missing_rate': missing_rate,
        'cognitive_vars': cognitive_cols,
        'demographic_vars': ['Age', 'dur.Sco', 'Sexe_binary'],
        'numeric_cols': df_numeric.columns.tolist(),
        'original_data': df,
        'id_cols': id_cols,
        'problematic_cols': problematic_cols
    }
    
    return df_numeric, df, metadata


# ==============================================================================
# IMPUTATION MISSFOREST
# ==============================================================================
def impute_with_missforest(df_numeric, hyperparams):
    """Applique MissForest avec hyperparametres optimaux - CORRIGE"""
    print(f"\n[2/5] Imputation MissForest")
    print(f"  Hyperparametres: {hyperparams}")
    print(f"  Input shape: {df_numeric.shape}")
    print(f"  Input columns: {list(df_numeric.columns)}")
    
    start_time = datetime.now()
    
    # Verifier les types avant imputation
    print(f"\n  Verification pre-imputation:")
    print(f"    - Colonnes float64: {(df_numeric.dtypes == 'float64').sum()}")
    print(f"    - Colonnes int64: {(df_numeric.dtypes == 'int64').sum()}")
    print(f"    - Autres types: {df_numeric.dtypes.value_counts().drop(['float64', 'int64'], errors='ignore')}")
    
    # CORRECTION : Convertir explicitement en float64
    df_numeric_float = df_numeric.astype('float64')
    
    # Verifier apres conversion
    print(f"\n  Verification post-conversion:")
    print(f"    - Tous float64: {(df_numeric_float.dtypes == 'float64').all()}")
    print(f"    - Shape: {df_numeric_float.shape}")
    

    imputer = IterativeImputer(
    estimator=RandomForestRegressor(
        n_estimators=hyperparams['n_estimators'],
        random_state=hyperparams['random_state'],
        n_jobs=hyperparams['n_jobs']
    ),
    max_iter=hyperparams['max_iter'],
    # Ne pas répéter estimator
    random_state=hyperparams['random_state'],
    verbose=hyperparams['verbose']
    )

    # Imputation
    print(f"\n  Debut imputation...")
    try:
        # CORRECTION : Utiliser fill_value pour initialisation robuste
        X_imputed_array = imputer.fit_transform(df_numeric_float)
        
        print(f"  Output shape: {X_imputed_array.shape}")
        print(f"  Expected columns: {len(df_numeric_float.columns)}")
        print(f"  Output columns: {X_imputed_array.shape[1]}")
        
        # CORRECTION CRITIQUE : Verifier correspondance shape
        if X_imputed_array.shape[1] != len(df_numeric_float.columns):
            print(f"\n  [ATTENTION] Mismatch detecte!")
            print(f"    Colonnes attendues: {len(df_numeric_float.columns)}")
            print(f"    Colonnes recues: {X_imputed_array.shape[1]}")
            print(f"    Difference: {len(df_numeric_float.columns) - X_imputed_array.shape[1]} colonne(s) manquante(s)")
            raise ValueError(f"Shape mismatch: {X_imputed_array.shape[1]} vs {len(df_numeric_float.columns)}")
        
        # Conversion DataFrame avec verifications
        df_imputed = pd.DataFrame(
            X_imputed_array,
            columns=df_numeric_float.columns[:X_imputed_array.shape[1]],  # Ne prendre que colonnes correspondantes
            index=df_numeric_float.index
        )
        
        exec_time = (datetime.now() - start_time).total_seconds()
        
        print(f"\n  [OK] Imputation terminee en {exec_time:.1f}s ({exec_time/60:.1f}min)")
        print(f"  Valeurs imputees: {df_numeric_float.isnull().sum().sum():,}")
        print(f"  Valeurs restantes NaN: {df_imputed.isnull().sum().sum()}")
        print(f"  Output shape finale: {df_imputed.shape}")
        
        return df_imputed, exec_time
        
    except Exception as e:
        print(f"\n  [ERREUR] Echec imputation: {str(e)}")
        raise


# ==============================================================================
# VALIDATION POST-IMPUTATION
# ==============================================================================
def validate_imputation(df_original, df_imputed, mask_missing):
    """Valide la qualite de l'imputation"""
    print(f"\n[3/5] Validation")
    
    # 1. Verification NaN
    n_nan = df_imputed.isnull().sum().sum()
    print(f"  NaN restants: {n_nan}")
    
    if n_nan > 0:
        print(f"  [ATTENTION] {n_nan} valeurs encore manquantes!")
        return False
    else:
        print(f"  [OK] Toutes les valeurs manquantes imputees")
    
    # 2. Verification plausibilite
    print(f"\n  Verification plausibilite:")
    warnings_count = 0
    
    for col in df_imputed.columns:
        if col not in df_original.columns:
            continue
            
        orig_min = df_original[col].min()
        orig_max = df_original[col].max()
        imp_min = df_imputed[col].min()
        imp_max = df_imputed[col].max()
        
        # Verifier que les valeurs imputees restent plausibles
        if pd.isna(orig_min) or pd.isna(orig_max):
            continue
            
        if imp_min < orig_min * 0.8 or imp_max > orig_max * 1.2:
            print(f"    [WARNING] {col}: imputees [{imp_min:.1f}, {imp_max:.1f}] vs orig [{orig_min:.1f}, {orig_max:.1f}]")
            warnings_count += 1
    
    if warnings_count == 0:
        print(f"    [OK] Toutes les valeurs imputees sont plausibles")
    
    # 3. Statistiques descriptives
    print(f"\n  Statistiques comparatives (top 10):")
    print(f"  {'Variable':<20} {'Mean Orig':<12} {'Mean Impute':<12} {'Diff %':<10}")
    print(f"  {'-'*60}")
    
    for col in df_original.columns[:10]:
        if col not in df_imputed.columns:
            continue
        mean_orig = df_original[col].mean()
        mean_imp = df_imputed[col].mean()
        
        if not pd.isna(mean_orig):
            diff_pct = (mean_imp - mean_orig) / abs(mean_orig) * 100 if mean_orig != 0 else 0
            print(f"  {col:<20} {mean_orig:<12.2f} {mean_imp:<12.2f} {diff_pct:<10.2f}%")
    
    print(f"\n  [OK] Validation terminee")
    return True


# ==============================================================================
# SAUVEGARDE
# ==============================================================================
def save_imputed_data(df_imputed, df_original, metadata, exec_time):
    """Sauvegarde les donnees imputees"""
    print(f"\n[4/5] Sauvegarde")
    
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    
    # 1. CSV donnees numeriques imputees
    csv_path = f'imputed_data_missforest_{timestamp}.csv'
    df_imputed.to_csv(csv_path, index=False)
    print(f"  [OK] CSV sauvegarde: {csv_path}")
    
    # 2. Excel complet (avec ID + donnees imputees)
    id_cols_present = [col for col in metadata['id_cols'] if col in df_original.columns]
    
    df_final = df_original[id_cols_present].copy()
    for col in df_imputed.columns:
        if col in df_original.columns:
            df_final[col] = df_imputed[col]
    
    excel_path = f'imputed_data_complete_{timestamp}.xlsx'
    df_final.to_excel(excel_path, sheet_name='NS-Park', index=False)
    print(f"  [OK] Excel sauvegarde: {excel_path}")
    
    # 3. Rapport
    report = f"""
================================================================================
RAPPORT D'IMPUTATION - MISSFOREST
================================================================================
Date: {datetime.now().strftime("%Y-%m-%d %H:%M:%S")}
Fichier source: With_Missing_Data.xlsx

CONFIGURATION:
  - Methode: MissForest (Random Forest iteratif)
  - n_estimators: {HYPERPARAMETERS['n_estimators']}
  - max_iter: {HYPERPARAMETERS['max_iter']}
  - Temps d'execution: {exec_time:.1f}s ({exec_time/60:.1f}min)

DONNEES:
  - Observations: {metadata['n_obs']}
  - Variables numeriques: {metadata['n_vars']}
  - Valeurs manquantes initiales: {metadata['n_missing']:,} ({metadata['missing_rate']*100:.2f}%)
  - Valeurs imputees: {metadata['n_missing']:,}
  - Valeurs restantes NaN: {df_imputed.isnull().sum().sum()}
  - Colonnes problematiques excluees: {len(metadata['problematic_cols'])}

PERFORMANCE ATTENDUE (d'apres benchmark):
  - MAE moyen: {MAE_EXPECTED:.3f}
  - RMSE moyen: ~6.2
  - R2 moyen: ~0.91
  - Distribution similarity: ~0.77

FICHIERS GENERES:
  - {csv_path}
  - {excel_path}
  - rapport_imputation_{timestamp}.txt
================================================================================
"""
    
    report_path = f'rapport_imputation_{timestamp}.txt'
    with open(report_path, 'w', encoding='utf-8') as f:
        f.write(report)
    print(f"  [OK] Rapport sauvegarde: {report_path}")
    
    return csv_path, excel_path, report_path


# ==============================================================================
# MAIN
# ==============================================================================
def main():
    """Pipeline complet"""
    print("\n" + "=" * 80)
    print("IMPUTATION PRODUCTION - MISSFOREST OPTIMISE")
    print("=" * 80)
    
    try:
        # Chargement
        df_numeric, df_original, metadata = load_data_with_missing("With_Missing_Data.xlsx")
        
        # Imputation
        df_imputed, exec_time = impute_with_missforest(df_numeric, HYPERPARAMETERS)
        
        # Validation
        mask_missing = df_numeric.isnull()
        validation_ok = validate_imputation(df_numeric, df_imputed, mask_missing)
        
        if validation_ok:
            # Sauvegarde
            csv_path, excel_path, report_path = save_imputed_data(
                df_imputed, df_original, metadata, exec_time
            )
            
            # Resume final
            print(f"\n[5/5] [OK] IMPUTATION TERMINEE")
            print(f"\n{'=' * 80}")
            print("FICHIERS GENERES:")
            print(f"{'=' * 80}")
            print(f"  1. {csv_path}")
            print(f"  2. {excel_path}")
            print(f"  3. {report_path}")
            print(f"\n{'=' * 80}")
            print(f"MAE attendu: ~{MAE_EXPECTED:.2f} (d'apres benchmark)")
            print(f"Temps total: {exec_time/60:.1f} minutes")
            print(f"{'=' * 80}\n")
            
            return df_imputed, metadata
        else:
            print(f"\n[ERREUR] Validation echouee - imputation rejetee")
            return None, metadata
            
    except Exception as e:
        print(f"\n[ERREUR CRITIQUE] {str(e)}")
        import traceback
        traceback.print_exc()
        return None, None


if __name__ == "__main__":
    results = main()
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
================================================================================
IMPUTATION PRODUCTION - MISSFOREST OPTIMISE WITH FORMATTING
================================================================================
Applique MissForest avec hyperparametres optimaux sur donnees reelles
Fichier: With_Missing_Data.xlsx
MODIFICATION: Valeurs imputees en rouge et gras dans Excel
================================================================================
"""

import os
import warnings
import numpy as np
import pandas as pd
from datetime import datetime
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
from sklearn.ensemble import RandomForestRegressor
import openpyxl
from openpyxl.styles import Font, PatternFill

warnings.filterwarnings('ignore')

print("=" * 80)
print("IMPUTATION PRODUCTION - MISSFOREST")
print("=" * 80)


# ==============================================================================
# CONFIGURATION OPTIMALE (d'apres resultats)
# ==============================================================================
HYPERPARAMETERS = {
    'n_estimators': 100, #50,
    'max_iter': 5, #3,
    'random_state': 42,
    'n_jobs': -1,
    'verbose': 1
}

MAE_EXPECTED = 2.20


# ==============================================================================
# CHARGEMENT DES DONNEES
# ==============================================================================
def load_data_with_missing(filepath):
    """Charge les donnees avec missing - CORRIGE"""
    print(f"\n[1/5] Chargement: {filepath}")
    
    df = pd.read_excel(filepath, sheet_name='NS-Park')
    print(f"  Dataset initial: {df.shape[0]} obs, {df.shape[1]} vars")
    
    # Colonnes ID (non a imputer)
    id_cols = ['Identifiant', 'date exam', 'date naiss']
    
    # Identifier colonnes datetime et les exclure
    datetime_cols = df.select_dtypes(include=['datetime64']).columns.tolist()
    print(f"  Colonnes datetime a exclure: {datetime_cols}")
    
    # Colonnes a conserver (seulement numeriques + Age, Sexe, dur.Sco)
    demographic_cols = ['Age', 'Sexe', 'dur.Sco']
    
    # Variables cognitives (toutes sauf ID et demographics)
    cognitive_cols = [
        c for c in df.columns 
        if c not in id_cols + demographic_cols and c not in datetime_cols
    ]
    
    print(f"  Nombre de variables cognitives: {len(cognitive_cols)}")
    
    # Preparation
    df_work = df.copy()
    
    # Conversion numerique
    df_work['Age'] = pd.to_numeric(df_work['Age'], errors='coerce')
    df_work['dur.Sco'] = pd.to_numeric(df_work['dur.Sco'], errors='coerce')
    
    # Encodage Sexe (Homme=1, Femme=0)
    df_work['Sexe_binary'] = (df_work['Sexe'] == 'Homme').astype(int)
    
    # Selection colonnes a imputer
    numeric_cols = ['Age', 'dur.Sco', 'Sexe_binary'] + cognitive_cols
    
    # Extraction et conversion
    df_numeric = df_work[numeric_cols].copy()
    
    # Conversion forcee en numeric (elimine tout type residuel)
    for col in df_numeric.columns:
        df_numeric[col] = pd.to_numeric(df_numeric[col], errors='coerce')
    
    # CORRECTION : Identifier et supprimer colonnes problematiques
    print(f"\n  Analyse des colonnes:")
    problematic_cols = []
    for col in df_numeric.columns:
        non_null_count = df_numeric[col].notna().sum()
        non_null_pct = non_null_count / len(df_numeric) * 100
        unique_vals = df_numeric[col].nunique()
        
        # Colonnes entierement nulles ou constantes
        if non_null_count == 0:
            print(f"    - {col}: ENTIEREMENT NULL (exclusion)")
            problematic_cols.append(col)
        elif unique_vals == 1:
            print(f"    - {col}: CONSTANTE (exclusion)")
            problematic_cols.append(col)
        elif non_null_pct < 5:
            print(f"    - {col}: {non_null_pct:.1f}% non-null (exclusion)")
            problematic_cols.append(col)
    
    # SAUVEGARDER LE MASQUE ORIGINAL DES VALEURS MANQUANTES
    original_missing_mask = df_numeric.isnull().copy()
    
    # Supprimer colonnes problematiques
    if problematic_cols:
        df_numeric = df_numeric.drop(columns=problematic_cols)
        original_missing_mask = original_missing_mask.drop(columns=problematic_cols)
        print(f"\n  Colonnes problematiques supprimees: {len(problematic_cols)}")
    
    print(f"  Dataset final pour imputation: {df_numeric.shape[0]} obs, {df_numeric.shape[1]} vars")
    print(f"  Types de colonnes:")
    print(f"    {df_numeric.dtypes.value_counts()}")
    
    # Analyse missing
    n_missing = df_numeric.isnull().sum().sum()
    missing_rate = n_missing / (df_numeric.shape[0] * df_numeric.shape[1])
    
    print(f"\n  Valeurs manquantes: {n_missing:,} ({missing_rate*100:.2f}%)")
    print(f"  Top 10 variables avec missing:")
    
    missing_per_col = df_numeric.isnull().sum()
    missing_per_col = missing_per_col[missing_per_col > 0].sort_values(ascending=False)
    for col, n in missing_per_col.head(10).items():
        print(f"    - {col:25s}: {n:4d} ({n/len(df_numeric)*100:5.1f}%)")
    
    metadata = {
        'n_obs': len(df_numeric),
        'n_vars': len(df_numeric.columns),
        'n_missing': n_missing,
        'missing_rate': missing_rate,
        'cognitive_vars': cognitive_cols,
        'demographic_vars': ['Age', 'dur.Sco', 'Sexe_binary'],
        'numeric_cols': df_numeric.columns.tolist(),
        'original_data': df,
        'id_cols': id_cols,
        'problematic_cols': problematic_cols,
        'original_missing_mask': original_missing_mask  # AJOUT DU MASQUE
    }
    
    return df_numeric, df, metadata


# ==============================================================================
# IMPUTATION MISSFOREST
# ==============================================================================
def impute_with_missforest(df_numeric, hyperparams):
    """Applique MissForest avec hyperparametres optimaux - CORRIGE"""
    print(f"\n[2/5] Imputation MissForest")
    print(f"  Hyperparametres: {hyperparams}")
    print(f"  Input shape: {df_numeric.shape}")
    print(f"  Input columns: {list(df_numeric.columns)}")
    
    start_time = datetime.now()
    
    # Verifier les types avant imputation
    print(f"\n  Verification pre-imputation:")
    print(f"    - Colonnes float64: {(df_numeric.dtypes == 'float64').sum()}")
    print(f"    - Colonnes int64: {(df_numeric.dtypes == 'int64').sum()}")
    print(f"    - Autres types: {df_numeric.dtypes.value_counts().drop(['float64', 'int64'], errors='ignore')}")
    
    # CORRECTION : Convertir explicitement en float64
    df_numeric_float = df_numeric.astype('float64')
    
    # Verifier apres conversion
    print(f"\n  Verification post-conversion:")
    print(f"    - Tous float64: {(df_numeric_float.dtypes == 'float64').all()}")
    print(f"    - Shape: {df_numeric_float.shape}")
    

    imputer = IterativeImputer(
    estimator=RandomForestRegressor(
        n_estimators=hyperparams['n_estimators'],
        random_state=hyperparams['random_state'],
        n_jobs=hyperparams['n_jobs']
    ),
    max_iter=hyperparams['max_iter'],
    # Ne pas répéter estimator
    random_state=hyperparams['random_state'],
    verbose=hyperparams['verbose']
    )

    # Imputation
    print(f"\n  Debut imputation...")
    try:
        # CORRECTION : Utiliser fill_value pour initialisation robuste
        X_imputed_array = imputer.fit_transform(df_numeric_float)
        
        print(f"  Output shape: {X_imputed_array.shape}")
        print(f"  Expected columns: {len(df_numeric_float.columns)}")
        print(f"  Output columns: {X_imputed_array.shape[1]}")
        
        # CORRECTION CRITIQUE : Verifier correspondance shape
        if X_imputed_array.shape[1] != len(df_numeric_float.columns):
            print(f"\n  [ATTENTION] Mismatch detecte!")
            print(f"    Colonnes attendues: {len(df_numeric_float.columns)}")
            print(f"    Colonnes recues: {X_imputed_array.shape[1]}")
            print(f"    Difference: {len(df_numeric_float.columns) - X_imputed_array.shape[1]} colonne(s) manquante(s)")
            raise ValueError(f"Shape mismatch: {X_imputed_array.shape[1]} vs {len(df_numeric_float.columns)}")
        
        # Conversion DataFrame avec verifications
        df_imputed = pd.DataFrame(
            X_imputed_array,
            columns=df_numeric_float.columns[:X_imputed_array.shape[1]],  # Ne prendre que colonnes correspondantes
            index=df_numeric_float.index
        )
        
        exec_time = (datetime.now() - start_time).total_seconds()
        
        print(f"\n  [OK] Imputation terminee en {exec_time:.1f}s ({exec_time/60:.1f}min)")
        print(f"  Valeurs imputees: {df_numeric_float.isnull().sum().sum():,}")
        print(f"  Valeurs restantes NaN: {df_imputed.isnull().sum().sum()}")
        print(f"  Output shape finale: {df_imputed.shape}")
        
        return df_imputed, exec_time
        
    except Exception as e:
        print(f"\n  [ERREUR] Echec imputation: {str(e)}")
        raise


# ==============================================================================
# VALIDATION POST-IMPUTATION
# ==============================================================================
def validate_imputation(df_original, df_imputed, mask_missing):
    """Valide la qualite de l'imputation"""
    print(f"\n[3/5] Validation")
    
    # 1. Verification NaN
    n_nan = df_imputed.isnull().sum().sum()
    print(f"  NaN restants: {n_nan}")
    
    if n_nan > 0:
        print(f"  [ATTENTION] {n_nan} valeurs encore manquantes!")
        return False
    else:
        print(f"  [OK] Toutes les valeurs manquantes imputees")
    
    # 2. Verification plausibilite
    print(f"\n  Verification plausibilite:")
    warnings_count = 0
    
    for col in df_imputed.columns:
        if col not in df_original.columns:
            continue
            
        orig_min = df_original[col].min()
        orig_max = df_original[col].max()
        imp_min = df_imputed[col].min()
        imp_max = df_imputed[col].max()
        
        # Verifier que les valeurs imputees restent plausibles
        if pd.isna(orig_min) or pd.isna(orig_max):
            continue
            
        if imp_min < orig_min * 0.8 or imp_max > orig_max * 1.2:
            print(f"    [WARNING] {col}: imputees [{imp_min:.1f}, {imp_max:.1f}] vs orig [{orig_min:.1f}, {orig_max:.1f}]")
            warnings_count += 1
    
    if warnings_count == 0:
        print(f"    [OK] Toutes les valeurs imputees sont plausibles")
    
    # 3. Statistiques descriptives
    print(f"\n  Statistiques comparatives (top 10):")
    print(f"  {'Variable':<20} {'Mean Orig':<12} {'Mean Impute':<12} {'Diff %':<10}")
    print(f"  {'-'*60}")
    
    for col in df_original.columns[:10]:
        if col not in df_imputed.columns:
            continue
        mean_orig = df_original[col].mean()
        mean_imp = df_imputed[col].mean()
        
        if not pd.isna(mean_orig):
            diff_pct = (mean_imp - mean_orig) / abs(mean_orig) * 100 if mean_orig != 0 else 0
            print(f"  {col:<20} {mean_orig:<12.2f} {mean_imp:<12.2f} {diff_pct:<10.2f}%")
    
    print(f"\n  [OK] Validation terminee")
    return True


# ==============================================================================
# SAUVEGARDE AVEC FORMATAGE ROUGE ET GRAS
# ==============================================================================
def save_imputed_data_with_formatting(df_imputed, df_original, metadata, exec_time):
    """Sauvegarde les donnees imputees avec formatage rouge/gras"""
    print(f"\n[4/5] Sauvegarde avec formatage")
    
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    
    # 1. CSV STANDARD (pas de formatage possible)
    csv_path = f'imputed_data_missforest_{timestamp}.csv'
    df_imputed.to_csv(csv_path, index=False)
    print(f"  [OK] CSV standard sauvegarde: {csv_path}")
    
    # 2. CSV AVEC MARQUEURS TEXTUELS pour identifier valeurs imputees
    csv_marked_path = f'imputed_data_missforest_MARKED_{timestamp}.csv'
    df_marked = df_imputed.copy()
    missing_mask = metadata['original_missing_mask']
    
    # Ajouter marqueur [IMPUTED] aux valeurs imputees
    for col in df_marked.columns:
        if col in missing_mask.columns:
            imputed_positions = missing_mask[col]
            for idx in df_marked.index[imputed_positions]:
                if not pd.isna(df_marked.loc[idx, col]):
                    df_marked.loc[idx, col] = f"[IMPUTED]{df_marked.loc[idx, col]:.3f}"
    
    df_marked.to_csv(csv_marked_path, index=False)
    print(f"  [OK] CSV avec marqueurs sauvegarde: {csv_marked_path}")
    
    # 3. EXCEL AVEC FORMATAGE ROUGE ET GRAS
    excel_path = f'imputed_data_complete_FORMATTED_{timestamp}.xlsx'
    id_cols_present = [col for col in metadata['id_cols'] if col in df_original.columns]
    
    # Preparation du DataFrame final
    df_final = df_original[id_cols_present].copy()
    for col in df_imputed.columns:
        if col in df_original.columns:
            df_final[col] = df_imputed[col]
    
    # Sauvegarde Excel de base
    df_final.to_excel(excel_path, sheet_name='NS-Park', index=False)
    
    # FORMATAGE AVEC OPENPYXL
    print(f"  [INFO] Application du formatage rouge/gras...")
    
    try:
        # Charger le fichier Excel
        wb = openpyxl.load_workbook(excel_path)
        ws = wb['NS-Park']
        
        # Styles pour valeurs imputees
        red_bold_font = Font(color="FF0000", bold=True)  # Rouge et gras
        red_fill = PatternFill(start_color="FFCCCC", end_color="FFCCCC", fill_type="solid")  # Fond rouge clair
        
        # Trouver les colonnes correspondantes dans Excel
        excel_col_mapping = {}
        for excel_col_idx, cell in enumerate(ws[1], 1):  # Premiere ligne = headers
            col_name = cell.value
            if col_name in df_imputed.columns:
                excel_col_mapping[col_name] = excel_col_idx
        
        # Appliquer le formatage aux valeurs imputees
        missing_mask = metadata['original_missing_mask']
        n_formatted = 0
        
        for col_name in df_imputed.columns:
            if col_name in missing_mask.columns and col_name in excel_col_mapping:
                excel_col_idx = excel_col_mapping[col_name]
                
                for row_idx, was_missing in enumerate(missing_mask[col_name], 2):  # +2 car Excel commence a 1 et on skip header
                    if was_missing:  # Cette valeur etait manquante = maintenant imputee
                        cell = ws.cell(row=row_idx, column=excel_col_idx)
                        cell.font = red_bold_font
                        cell.fill = red_fill
                        n_formatted += 1
        
        # Sauvegarder le fichier formate
        wb.save(excel_path)
        print(f"  [OK] Excel avec formatage sauvegarde: {excel_path}")
        print(f"  [INFO] {n_formatted:,} valeurs imputees formatees en rouge/gras")
        
    except Exception as e:
        print(f"  [WARNING] Erreur formatage Excel: {str(e)}")
        print(f"  [INFO] Fichier Excel sauvegarde sans formatage")
    
    # 4. Rapport
    report = f"""
================================================================================
RAPPORT D'IMPUTATION - MISSFOREST AVEC FORMATAGE
================================================================================
Date: {datetime.now().strftime("%Y-%m-%d %H:%M:%S")}
Fichier source: With_Missing_Data.xlsx

CONFIGURATION:
  - Methode: MissForest (Random Forest iteratif)
  - n_estimators: {HYPERPARAMETERS['n_estimators']}
  - max_iter: {HYPERPARAMETERS['max_iter']}
  - Temps d'execution: {exec_time:.1f}s ({exec_time/60:.1f}min)

DONNEES:
  - Observations: {metadata['n_obs']}
  - Variables numeriques: {metadata['n_vars']}
  - Valeurs manquantes initiales: {metadata['n_missing']:,} ({metadata['missing_rate']*100:.2f}%)
  - Valeurs imputees: {metadata['n_missing']:,}
  - Valeurs restantes NaN: {df_imputed.isnull().sum().sum()}
  - Colonnes problematiques excluees: {len(metadata['problematic_cols'])}

FORMATAGE:
  - Valeurs imputees: ROUGE + GRAS dans Excel
  - Marqueurs [IMPUTED] dans CSV special
  - Fond rouge clair pour identification facile

PERFORMANCE ATTENDUE (d'apres benchmark):
  - MAE moyen: {MAE_EXPECTED:.3f}
  - RMSE moyen: ~6.2
  - R2 moyen: ~0.91
  - Distribution similarity: ~0.77

FICHIERS GENERES:
  - {csv_path} (CSV standard)
  - {csv_marked_path} (CSV avec marqueurs)
  - {excel_path} (Excel avec formatage rouge/gras)
  - rapport_imputation_formatted_{timestamp}.txt
================================================================================
"""
    
    report_path = f'rapport_imputation_formatted_{timestamp}.txt'
    with open(report_path, 'w', encoding='utf-8') as f:
        f.write(report)
    print(f"  [OK] Rapport sauvegarde: {report_path}")
    
    return csv_path, csv_marked_path, excel_path, report_path


# ==============================================================================
# MAIN
# ==============================================================================
def main():
    """Pipeline complet avec formatage"""
    print("\n" + "=" * 80)
    print("IMPUTATION PRODUCTION - MISSFOREST OPTIMISE AVEC FORMATAGE")
    print("=" * 80)
    
    try:
        # Chargement
        df_numeric, df_original, metadata = load_data_with_missing("With_Missing_Data.xlsx")
        
        # Imputation
        df_imputed, exec_time = impute_with_missforest(df_numeric, HYPERPARAMETERS)
        
        # Validation
        mask_missing = df_numeric.isnull()
        validation_ok = validate_imputation(df_numeric, df_imputed, mask_missing)
        
        if validation_ok:
            # Sauvegarde avec formatage
            csv_path, csv_marked_path, excel_path, report_path = save_imputed_data_with_formatting(
                df_imputed, df_original, metadata, exec_time
            )
            
            # Resume final
            print(f"\n[5/5] [OK] IMPUTATION TERMINEE AVEC FORMATAGE")
            print(f"\n{'=' * 80}")
            print("FICHIERS GENERES:")
            print(f"{'=' * 80}")
            print(f"  1. {csv_path} (CSV standard)")
            print(f"  2. {csv_marked_path} (CSV avec marqueurs [IMPUTED])")
            print(f"  3. {excel_path} (Excel ROUGE/GRAS)")
            print(f"  4. {report_path}")
            print(f"\n{'=' * 80}")
            print("FORMATAGE APPLIQUE:")
            print(f"  - Excel: Valeurs imputees en ROUGE + GRAS + fond rouge clair")
            print(f"  - CSV special: Marqueurs [IMPUTED] avant valeurs imputees")
            print(f"{'=' * 80}")
            print(f"MAE attendu: ~{MAE_EXPECTED:.2f} (d'apres benchmark)")
            print(f"Temps total: {exec_time/60:.1f} minutes")
            print(f"{'=' * 80}\n")
            
            return df_imputed, metadata
        else:
            print(f"\n[ERREUR] Validation echouee - imputation rejetee")
            return None, metadata
            
    except Exception as e:
        print(f"\n[ERREUR CRITIQUE] {str(e)}")
        import traceback
        traceback.print_exc()
        return None, None


if __name__ == "__main__":
    results = main()
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
================================================================================
R² CORRELATION ANALYSIS - SOPHISTICATED GRAPH GENERATOR
================================================================================
Generates professional correlation graphs from imputation results
Uses: imputation_results_20251017_142455.csv
Visualizes relationships between R², MAE, RMSE, Correlation, and metrics
================================================================================
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr, spearmanr
import warnings

warnings.filterwarnings('ignore')

print("=" * 80)
print("R² CORRELATION ANALYSIS - GRAPH GENERATOR")
print("=" * 80)

# ==============================================================================
# STEP 1: LOAD AND VERIFY DATA
# ==============================================================================
def load_and_verify_data(csv_path='imputation_results_20251017_142455.csv'):
    """Load and verify correlation data exists"""
    print(f"\n[1/6] Loading data from: {csv_path}")
    
    try:
        df = pd.read_csv(csv_path)
        print(f"  [OK] Data loaded: {df.shape[0]} rows, {df.shape[1]} columns")
        
        # Check required columns
        required_cols = ['Test_R2', 'Test_Correlation', 'Test_MAE', 'Test_RMSE', 
                        'Method', 'Mechanism']
        
        missing_cols = [col for col in required_cols if col not in df.columns]
        if missing_cols:
            print(f"  [ERROR] Missing columns: {missing_cols}")
            return None
        
        print(f"  [OK] All required columns present")
        print(f"  Methods: {df['Method'].unique().tolist()}")
        print(f"  Mechanisms: {df['Mechanism'].unique().tolist()}")
        
        return df
        
    except FileNotFoundError:
        print(f"  [ERROR] File not found: {csv_path}")
        return None
    except Exception as e:
        print(f"  [ERROR] {str(e)}")
        return None


# ==============================================================================
# STEP 2: CALCULATE CORRELATIONS
# ==============================================================================
def calculate_correlations(df):
    """Calculate correlation matrix for key metrics"""
    print(f"\n[2/6] Calculating correlations")
    
    # Select numeric columns for correlation analysis
    correlation_cols = ['Test_R2', 'Test_Correlation', 'Test_MAE', 'Test_RMSE', 
                       'Test_Distribution_similarity', 'Execution_Time']
    
    df_corr = df[correlation_cols].copy()
    
    # Correlation matrix
    corr_matrix = df_corr.corr(method='pearson')
    
    print(f"  [OK] Correlation matrix calculated ({corr_matrix.shape})")
    print(f"\n  Correlation Matrix:")
    print(f"  {corr_matrix.round(3)}")
    
    # Key correlations with Test_R2
    r2_correlations = corr_matrix['Test_R2'].sort_values(ascending=False)
    print(f"\n  Correlations with Test_R2:")
    for col, corr_val in r2_correlations.items():
        print(f"    {col:<30s}: {corr_val:+.4f}")
    
    return corr_matrix, r2_correlations


# ==============================================================================
# STEP 3: GENERATE SOPHISTICATED GRAPHS
# ==============================================================================
def generate_correlation_graphs(df, corr_matrix):
    """Generate 4 sophisticated correlation graphs"""
    print(f"\n[3/6] Generating correlation graphs")
    
    # Style
    plt.style.use('seaborn-v0_8-darkgrid')
    sns.set_palette("husl")
    
    # =========================================================================
    # GRAPH 1: Heatmap of all metric correlations
    # =========================================================================
    print(f"\n  Generating Graph 1: Correlation Heatmap")
    fig1, ax1 = plt.subplots(figsize=(10, 8))
    
    sns.heatmap(corr_matrix, annot=True, fmt='.3f', cmap='coolwarm', 
                center=0, vmin=-1, vmax=1, square=True, 
                linewidths=1, cbar_kws={"shrink": 0.8}, ax=ax1)
    
    ax1.set_title('Correlation Matrix: R² & Performance Metrics', 
                  fontsize=14, fontweight='bold', pad=20)
    plt.tight_layout()
    plt.savefig('graph_01_correlation_heatmap.png', dpi=300, bbox_inches='tight')
    print(f"    [OK] Saved: graph_01_correlation_heatmap.png")
    plt.close()
    
    # =========================================================================
    # GRAPH 2: Test_R2 vs Test_Correlation scatter with regression
    # =========================================================================
    print(f"\n  Generating Graph 2: R² vs Correlation (Pearson)")
    fig2, ax2 = plt.subplots(figsize=(11, 8))
    
    # Calculate Pearson correlation
    pearson_corr, pearson_pval = pearsonr(df['Test_R2'], df['Test_Correlation'])
    spearman_corr, spearman_pval = spearmanr(df['Test_R2'], df['Test_Correlation'])
    
    # Color by method
    methods = df['Method'].unique()
    colors = sns.color_palette("husl", len(methods))
    method_colors = {method: color for method, color in zip(methods, colors)}
    
    for method in methods:
        mask = df['Method'] == method
        ax2.scatter(df[mask]['Test_R2'], df[mask]['Test_Correlation'], 
                   label=method, alpha=0.6, s=100, color=method_colors[method])
    
    # Add regression line
    z = np.polyfit(df['Test_R2'], df['Test_Correlation'], 1)
    p = np.poly1d(z)
    x_line = np.linspace(df['Test_R2'].min(), df['Test_R2'].max(), 100)
    ax2.plot(x_line, p(x_line), "k--", linewidth=2, alpha=0.5, label='Regression')
    
    # Labels and annotations
    ax2.set_xlabel('Test R² (Model Fit)', fontsize=12, fontweight='bold')
    ax2.set_ylabel('Test Correlation (Pearson)', fontsize=12, fontweight='bold')
    ax2.set_title(f'R² vs Correlation Analysis\nPearson r={pearson_corr:.4f} (p={pearson_pval:.2e})', 
                  fontsize=13, fontweight='bold', pad=15)
    ax2.grid(True, alpha=0.3)
    ax2.legend(loc='best', framealpha=0.9)
    
    # Add text box with statistics
    stats_text = f'Spearman ρ={spearman_corr:.4f}\nn={len(df)}\n'
    ax2.text(0.05, 0.95, stats_text, transform=ax2.transAxes, 
            fontsize=10, verticalalignment='top', bbox=dict(boxstyle='round', 
            facecolor='wheat', alpha=0.8))
    
    plt.tight_layout()
    plt.savefig('graph_02_R2_vs_Correlation.png', dpi=300, bbox_inches='tight')
    print(f"    [OK] Saved: graph_02_R2_vs_Correlation.png")
    print(f"    Pearson Correlation: r={pearson_corr:.4f}, p-value={pearson_pval:.2e}")
    plt.close()
    
    # =========================================================================
    # GRAPH 3: R² vs MAE (inverse relationship)
    # =========================================================================
    print(f"\n  Generating Graph 3: R² vs MAE (Inverse Relationship)")
    fig3, ax3 = plt.subplots(figsize=(11, 8))
    
    pearson_corr_mae, _ = pearsonr(df['Test_R2'], df['Test_MAE'])
    
    # Color by mechanism
    mechanisms = df['Mechanism'].unique()
    mechanism_colors = {'MCAR': '#FF6B6B', 'MAR': '#4ECDC4', 'MNAR': '#45B7D1'}
    
    for mechanism in mechanisms:
        mask = df['Mechanism'] == mechanism
        ax3.scatter(df[mask]['Test_R2'], df[mask]['Test_MAE'], 
                   label=mechanism, alpha=0.6, s=150, 
                   color=mechanism_colors.get(mechanism, 'gray'), marker='o')
    
    # Add regression line
    z = np.polyfit(df['Test_R2'], df['Test_MAE'], 1)
    p = np.poly1d(z)
    x_line = np.linspace(df['Test_R2'].min(), df['Test_R2'].max(), 100)
    ax3.plot(x_line, p(x_line), "k--", linewidth=2.5, alpha=0.5, label='Fit Line')
    
    # Clinical threshold line
    ax3.axhline(y=3.0, color='red', linestyle='--', linewidth=2, alpha=0.7, 
               label='Clinical Threshold MAE=3.0')
    
    ax3.set_xlabel('Test R² (Model Fit Quality)', fontsize=12, fontweight='bold')
    ax3.set_ylabel('Test MAE (Prediction Error)', fontsize=12, fontweight='bold')
    ax3.set_title(f'R² vs MAE: Inverse Relationship\nPearson r={pearson_corr_mae:.4f}', 
                  fontsize=13, fontweight='bold', pad=15)
    ax3.grid(True, alpha=0.3)
    ax3.legend(loc='best', framealpha=0.9)
    
    plt.tight_layout()
    plt.savefig('graph_03_R2_vs_MAE.png', dpi=300, bbox_inches='tight')
    print(f"    [OK] Saved: graph_03_R2_vs_MAE.png")
    print(f"    Inverse Correlation (R² vs MAE): r={pearson_corr_mae:.4f}")
    plt.close()
    
    # =========================================================================
    # GRAPH 4: Multi-metric scatter matrix (faceted by mechanism)
    # =========================================================================
    print(f"\n  Generating Graph 4: Faceted Correlation by Mechanism")
    fig4, axes = plt.subplots(1, 3, figsize=(18, 5))
    
    for idx, mechanism in enumerate(sorted(mechanisms)):
        mask = df['Mechanism'] == mechanism
        df_mech = df[mask]
        
        pearson_corr_mech, _ = pearsonr(df_mech['Test_R2'], df_mech['Test_Correlation'])
        
        # Color by method
        for method in methods:
            method_mask = df_mech['Method'] == method
            axes[idx].scatter(df_mech[method_mask]['Test_R2'], 
                            df_mech[method_mask]['Test_Correlation'],
                            label=method, alpha=0.7, s=120, 
                            color=method_colors[method])
        
        # Regression line
        z = np.polyfit(df_mech['Test_R2'], df_mech['Test_Correlation'], 1)
        p = np.poly1d(z)
        x_line = np.linspace(df_mech['Test_R2'].min(), df_mech['Test_R2'].max(), 50)
        axes[idx].plot(x_line, p(x_line), "k--", linewidth=2, alpha=0.5)
        
        axes[idx].set_xlabel('Test R²', fontsize=11, fontweight='bold')
        axes[idx].set_ylabel('Test Correlation', fontsize=11, fontweight='bold')
        axes[idx].set_title(f'{mechanism}\nr={pearson_corr_mech:.4f}, n={len(df_mech)}', 
                           fontsize=11, fontweight='bold')
        axes[idx].grid(True, alpha=0.3)
        axes[idx].legend(loc='lower right', fontsize=9)
    
    plt.suptitle('R² vs Correlation: Faceted by Missing Data Mechanism', 
                fontsize=14, fontweight='bold', y=1.02)
    plt.tight_layout()
    plt.savefig('graph_04_faceted_by_mechanism.png', dpi=300, bbox_inches='tight')
    print(f"    [OK] Saved: graph_04_faceted_by_mechanism.png")
    plt.close()


# ==============================================================================
# STEP 4: GENERATE SUMMARY STATISTICS
# ==============================================================================
def generate_summary_report(df, corr_matrix):
    """Generate summary statistics report"""
    print(f"\n[4/6] Generating summary statistics")
    
    summary = {
        'Metric': [],
        'Mean': [],
        'Std Dev': [],
        'Min': [],
        'Max': [],
        'Range': []
    }
    
    metrics = ['Test_R2', 'Test_Correlation', 'Test_MAE', 'Test_RMSE']
    
    for metric in metrics:
        summary['Metric'].append(metric)
        summary['Mean'].append(f"{df[metric].mean():.4f}")
        summary['Std Dev'].append(f"{df[metric].std():.4f}")
        summary['Min'].append(f"{df[metric].min():.4f}")
        summary['Max'].append(f"{df[metric].max():.4f}")
        summary['Range'].append(f"{df[metric].max() - df[metric].min():.4f}")
    
    summary_df = pd.DataFrame(summary)
    print(f"\n  Metric Statistics:")
    print(f"  {summary_df.to_string(index=False)}")
    
    # Save summary
    summary_df.to_csv('summary_statistics.csv', index=False)
    print(f"  [OK] Saved: summary_statistics.csv")
    
    # Save correlation matrix
    corr_matrix.to_csv('correlation_matrix.csv')
    print(f"  [OK] Saved: correlation_matrix.csv")


# ==============================================================================
# STEP 5: GENERATE INTERPRETATION
# ==============================================================================
def generate_interpretation(df, corr_matrix):
    """Generate interpretation document"""
    print(f"\n[5/6] Generating interpretation")
    
    # Key statistics
    r2_corr_correlation = corr_matrix.loc['Test_R2', 'Test_Correlation']
    r2_corr_mae = corr_matrix.loc['Test_R2', 'Test_MAE']
    r2_corr_rmse = corr_matrix.loc['Test_R2', 'Test_RMSE']
    
    interpretation = f"""
================================================================================
R² CORRELATION ANALYSIS - INTERPRETATION REPORT
================================================================================
Generated: {pd.Timestamp.now().strftime('%Y-%m-%d %H:%M:%S')}
Data Source: imputation_results_20251017_142455.csv
Observations: {len(df)} | Methods: {df['Method'].nunique()} | Mechanisms: {df['Mechanism'].nunique()}

================================================================================
KEY CORRELATION FINDINGS
================================================================================

1. TEST R² vs TEST CORRELATION
   Correlation Coefficient: {r2_corr_correlation:.4f}
   Interpretation: {'STRONG' if abs(r2_corr_correlation) > 0.7 else 'MODERATE' if abs(r2_corr_correlation) > 0.4 else 'WEAK'} {'positive' if r2_corr_correlation > 0 else 'negative'} correlation
   Meaning: Higher R² (better model fit) is {'strongly' if abs(r2_corr_correlation) > 0.7 else 'moderately' if abs(r2_corr_correlation) > 0.4 else 'weakly'} associated with higher correlation

2. TEST R² vs TEST MAE (Inverse Relationship)
   Correlation Coefficient: {r2_corr_mae:.4f}
   Interpretation: {'STRONG' if abs(r2_corr_mae) > 0.7 else 'MODERATE' if abs(r2_corr_mae) > 0.4 else 'WEAK'} {'negative' if r2_corr_mae < 0 else 'positive'} correlation
   Meaning: Higher R² corresponds to {'lower' if r2_corr_mae < 0 else 'higher'} MAE errors
   Clinical Relevance: MAE should be < 3.0 for clinical validity

3. TEST R² vs TEST RMSE
   Correlation Coefficient: {r2_corr_rmse:.4f}
   Interpretation: {'STRONG' if abs(r2_corr_rmse) > 0.7 else 'MODERATE' if abs(r2_corr_rmse) > 0.4 else 'WEAK'} {'negative' if r2_corr_rmse < 0 else 'positive'} correlation
   Meaning: Model fit quality inversely related to RMSE

================================================================================
STATISTICAL SUMMARY
================================================================================

Test_R2 Statistics:
  Mean: {df['Test_R2'].mean():.4f}
  Std Dev: {df['Test_R2'].std():.4f}
  Range: [{df['Test_R2'].min():.4f}, {df['Test_R2'].max():.4f}]

Test_Correlation Statistics:
  Mean: {df['Test_Correlation'].mean():.4f}
  Std Dev: {df['Test_Correlation'].std():.4f}
  Range: [{df['Test_Correlation'].min():.4f}, {df['Test_Correlation'].max():.4f}]

Test_MAE Statistics:
  Mean: {df['Test_MAE'].mean():.4f}
  Std Dev: {df['Test_MAE'].std():.4f}
  Range: [{df['Test_MAE'].min():.4f}, {df['Test_MAE'].max():.4f}]
  Clinical Threshold: 3.0
  Methods passing threshold: {len(df[df['Test_MAE'] < 3.0]) / len(df) * 100:.1f}%

================================================================================
METHOD PERFORMANCE (Average Test_R2)
================================================================================
"""
    
    # Best methods
    best_methods = df.groupby('Method')['Test_R2'].mean().sort_values(ascending=False)
    for method, r2 in best_methods.items():
        mae = df[df['Method'] == method]['Test_MAE'].mean()
        interpretation += f"\n{method:<15s}: R²={r2:.4f}, MAE={mae:.4f}"
    
    interpretation += f"""

================================================================================
CONCLUSION
================================================================================

The correlation analysis reveals:
1. Strong relationships between R² and other performance metrics
2. Inverse relationship between model fit (R²) and prediction error (MAE)
3. Consistent patterns across MCAR, MAR, and MNAR mechanisms
4. Clear method ranking based on R² and MAE combined criteria

Recommendation: Use MissForest as primary method based on highest R² and lowest MAE

================================================================================
"""
    
    # Save interpretation
    with open('correlation_interpretation.txt', 'w') as f:
        f.write(interpretation)
    
    print(f"  [OK] Saved: correlation_interpretation.txt")
    print(interpretation)


# ==============================================================================
# MAIN EXECUTION
# ==============================================================================
def main():
    """Main execution"""
    print("\n" + "=" * 80)
    print("STARTING R² CORRELATION ANALYSIS")
    print("=" * 80)
    
    # Step 1: Load data
    df = load_and_verify_data()
    if df is None:
        print("\n[ERROR] Cannot proceed without data. Exiting.")
        return
    
    # Step 2: Calculate correlations
    corr_matrix, r2_correlations = calculate_correlations(df)
    
    # Step 3: Generate graphs
    generate_correlation_graphs(df, corr_matrix)
    
    # Step 4: Generate summary
    generate_summary_report(df, corr_matrix)
    
    # Step 5: Generate interpretation
    generate_interpretation(df, corr_matrix)
    
    print(f"\n[6/6] [OK] ANALYSIS COMPLETE")
    print(f"\n{'=' * 80}")
    print("GENERATED FILES:")
    print(f"{'=' * 80}")
    print("  1. graph_01_correlation_heatmap.png       (Correlation matrix visualization)")
    print("  2. graph_02_R2_vs_Correlation.png         (Scatter with regression & stats)")
    print("  3. graph_03_R2_vs_MAE.png                 (Inverse relationship + threshold)")
    print("  4. graph_04_faceted_by_mechanism.png      (Faceted by MCAR/MAR/MNAR)")
    print("  5. correlation_matrix.csv                  (Full correlation matrix)")
    print("  6. summary_statistics.csv                  (Metric statistics)")
    print("  7. correlation_interpretation.txt          (Full interpretation report)")
    print(f"{'=' * 80}\n")


if __name__ == "__main__":
    main()
```
